```python
#Ex 1 - Merge sort (sempre divide em dois até que exista apeans um)
def merge_sort(lista):
  if len(lista) <= 1:
    return lista

    meio = len(lista) // 2
    esquerdo = merge_sort(lista[:meio])
    direito = merge_sort(lista[meio:])

    resultado = []
    i = j = 0
    while i < len(esquerdo) and j < len(direito):
      if esquerdo[i] < direito[j]:
        resultado.append(esquerdo[i])
        i += 1
      else:
        resultado.append(direito[j])
        j += 1

    return resultado + esquerdo[i:] + direito[j:]

    lista_nao_ordenada = [5, 3, 8, 4, 2]
    print(merge_sort(lista_nao_ordenada))
```
