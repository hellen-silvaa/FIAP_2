AMOSTRAGEM

```python
# A amostragem é o processo de selecionar um subconjunto de elementos de uma população
# para realizar análises estatísticas ou modelagem. O objetivo é obter uma amostra
# que seja representativa da população, permitindo fazer inferências ou
# generalizações com maior eficiência e menor custo.

# Utilizações:
# * Pesquisas Eleitorais
# * Pesquisas Científicas
# * Testes e Gerenciamento de Memória

# Sorteio do Baralho (Naipes)
# Sorteio do Baralho (Cartas)
```

```python
#Numpy --------------------------------

#np.random.choice()

#Argumentos
#array: O array de onde os elementos serão amostrados.
#size: Número de amostras.
#replace: Define se a amostragem é com reposição (True) ou sem reposição (False).

#A função numpy.random.seed() é usada para inicializar o gerador de números aleatórios
# com um valor específico (a semente). Isso garante que os resultados dos métodos
# aleatórios sejam reprodutíveis.

#np.random.seed(x)

# --------------------------------

#Pandas --------------------------------
#df.sample()

#Argumentos
#n: Número de linhas a serem amostradas.
#frac: Fração do DataFrame a ser amostrada (0 a 1).
#replace: Amostragem com (True) ou sem (False) reposição.
#random_state: Para reprodutibilidade.
#axis: Escolhe entre amostragem de linhas (0) ou colunas (1).
```

```python
Dado = [1,2,3,4,5,6]
```

```python
print('Amostragem')
print(np.random.choice(Dado))
print('Múltipla Amostragem')
print(np.random.choice(Dado, size=3))
print('Amostragem sem Reposição')
print(np.random.choice(Dado, size=3, replace=False))
```

```python
# A função sample é usada para gerar uma amostra aleatória de itens de um DataFrame.

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item do conjunto.
print('Amostra aleatória de uma linha')
print(Data_Jobs.sample())

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente várias linhas do conjunto.
print('Amostra aleatória de mais de uma linha')
print(Data_Jobs.sample(4))
```

```python
# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de uma observação de uma coluna')
print(Data_Jobs['job_category'].sample())

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de uma observação de uma coluna')
print(Data_Jobs['job_category'].sample())

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de uma observação de uma coluna')
print(Data_Jobs['job_category'].sample())

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente 3 itens da coluna do conjunto.
print('Amostra aleatória de várias observações de uma coluna')
print(Data_Jobs['job_title'].sample(3))

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente 3 itens da coluna do conjunto.
print('Amostra aleatória de várias observações de uma coluna')
print(Data_Jobs['job_title'].sample(3))
```

```python
#O uso de random_state em um contexto de geração de amostras (ou em qualquer outro contexto que envolva a geração de números aleatórios)
#tem o propósito de tornar os resultados reproduzíveis. Quando você define uma semente (seed) para o gerador de números aleatórios,
#está essencialmente estabelecendo um ponto de partida fixo para a sequência de números que serão gerados.

print('Amostra aleatória de uma observação de uma coluna com random_state')
print(Data_Jobs['job_category'].sample(random_state=1))

print('-------------------------------------------------------')

print('Amostra aleatória de uma observação de uma coluna com random_state')
print(Data_Jobs['job_category'].sample(random_state=1))

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de várias observações de uma coluna com random_state')
print(Data_Jobs['job_category'].sample(3, random_state=2010))

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de várias observações de uma coluna com random_state')
print(Data_Jobs['job_category'].sample(3, random_state=2020))

print('-------------------------------------------------------')

# Exemplo de amostragem sem reposição: escolher aleatoriamente um item da coluna do conjunto.
print('Amostra aleatória de várias observações de uma coluna com random_state')
np.random.seed(2010)
print(Data_Jobs['job_category'].sample(3))
```

EVENTOS DEPENDENTES

```python
#Eventos são considerados dependentes se a ocorrência de um afeta a probabilidade de ocorrência do outro.
#Isso significa que o resultado de um evento tem um impacto sobre o resultado do outro.

#Cada cargo possui uma probabilidade no sorteio por amostragem
print('Probabilidade de cada cargo ser sorteado')
print(Data_Jobs_Func['job_title'].value_counts(normalize=True))

print('-------------------------------------------------------')

#Se no primeiro sorteio eu escolho um desses cargos, os demais vão ter uma probabilidade diferente
#de serem escolhidos no próximo sorteio
print('Sorteio aleatório de um cargo no dataframe')
print(Data_Jobs_Func['job_title'].sample(random_state=2))

print('-------------------------------------------------------')

#Retirando este cargo do dataframe
Data_Jobs_Func1 = Data_Jobs_Func[Data_Jobs_Func['job_title']!= 'AWS Data Architect']
print('Nova probabilidade dos cargos restantes serem sorteados após a retirada da primeira observação')
print(Data_Jobs_Func1['job_title'].value_counts(normalize=True))
```

EVENTOS INDEPENDENTES

```python
#Dois eventos são considerados independentes se a ocorrência de um não afeta a probabilidade de ocorrência do outro.
#Em outras palavras, o resultado de um evento não tem impacto sobre o resultado do outro.

#Cada cargo possui uma probabilidade no sorteio por amostragem
#Uma vez que, com o replacement a observação é adicionada novamente ao conjunto
#as linhas permanecem com a mesma probabilidade de serem sorteadas do que antes
print('Probabilidade de cada cargo ser sorteado')
print(Data_Jobs_Func['job_title'].value_counts(normalize=True))

print('-------------------------------------------------------')

#Cada cargo possui uma probabilidade no sorteio por amostragem
#Uma vez que, com o replacement a observação é adicionada novamente ao conjunto
#as linhas permanecem com a mesma probabilidade de serem sorteadas do que antes
print('Sorteio aleatório de um cargo no dataframe')
np.random.seed(2)
print(Data_Jobs_Func['job_title'].sample(4, replace=True))
```

```python
#A Lei dos Grandes Números (LLN) é um teorema central da probabilidade e estatística que descreve o resultado
#de realizar o mesmo experimento muitas vezes. De acordo com a LLN, à medida que o número de tentativas aumenta,
#a média dos resultados observados tende a se aproximar da expectativa matemática.

#Se jogarmos uma moeda justa um número grande de vezes, esperamos que o número de caras se aproxime de 50%
#do total de lançamentos. Mesmo que em poucos lançamentos a proporção possa variar significativamente
#(todos caras ou todos coroas), à medida que o número de lançamentos aumenta, a proporção tende a se estabilizar em torno de 50%.

#Importância Prática: Ajuda a explicar por que os cassinos sempre ganham a longo prazo e como as
#seguradoras estabelecem prêmios.

#Aplicação em Ciência de Dados: Fundamental para a validação de modelos estatísticos e para entender
#o comportamento de grandes conjuntos de dados.

print('Frequência de cada categoria no dataframe original')
print(Data_Jobs['job_category'].value_counts(normalize=True))

print('-------------------------------------------------------')

print('Frequência de cada categoria no dataframe com 100 observações')
print(Data_Jobs_S100.value_counts(normalize=True))
```

```python
DJ = Data_Jobs['job_category'].value_counts(normalize=True).reset_index()
DJ100 = Data_Jobs_S100.value_counts(normalize=True).reset_index()
DJ1000 = Data_Jobs_S1000.value_counts(normalize=True).reset_index()
DJ10000 = Data_Jobs_S10000.value_counts(normalize=True).reset_index()
DJ100000 = Data_Jobs_S100000.value_counts(normalize=True).reset_index()
DJ1000000 = Data_Jobs_S1000000.value_counts(normalize=True).reset_index()

DJ.columns=['job_category','prop']
DJ100.columns=['job_category','prop_100']
DJ1000.columns=['job_category','prop_1k']
DJ10000.columns=['job_category','prop_10k']
DJ100000.columns=['job_category','prop_100k']
DJ1000000.columns=['job_category','prop_1mm']
```

```python
Data_Jobs_Prob = pd.merge(DJ, DJ100, on='job_category', how='left')
Data_Jobs_Prob = pd.merge(Data_Jobs_Prob, DJ1000, on='job_category', how='left')
Data_Jobs_Prob = pd.merge(Data_Jobs_Prob, DJ10000, on='job_category', how='left')
Data_Jobs_Prob = pd.merge(Data_Jobs_Prob, DJ100000, on='job_category', how='left')
Data_Jobs_Prob = pd.merge(Data_Jobs_Prob, DJ1000000, on='job_category', how='left')
```

```python
Data_Jobs_Prob = Data_Jobs_Prob.fillna(0)
```

```python
Data_Jobs_Prob['Diff_100'] = np.abs(Data_Jobs_Prob['prop'] - Data_Jobs_Prob['prop_100'])
Data_Jobs_Prob['Diff_1k'] = np.abs(Data_Jobs_Prob['prop'] - Data_Jobs_Prob['prop_1k'])
Data_Jobs_Prob['Diff_10k'] = np.abs(Data_Jobs_Prob['prop'] - Data_Jobs_Prob['prop_10k'])
Data_Jobs_Prob['Diff_100k'] = np.abs(Data_Jobs_Prob['prop'] - Data_Jobs_Prob['prop_100k'])
Data_Jobs_Prob['Diff_1mm'] = np.abs(Data_Jobs_Prob['prop'] - Data_Jobs_Prob['prop_1mm'])
```

```python
print('Erro de estimativa com 100 obs')
print(Data_Jobs_Prob['Diff_100'].mean())

print('-------------------------------------------------------')

print('Erro de estimativa com 1k obs')
print(Data_Jobs_Prob['Diff_1k'].mean())

print('-------------------------------------------------------')

print('Erro de estimativa com 10k obs')
print(Data_Jobs_Prob['Diff_10k'].mean())

print('-------------------------------------------------------')

print('Erro de estimativa com 100k obs')
print(Data_Jobs_Prob['Diff_100k'].mean())

print('-------------------------------------------------------')

print('Erro de estimativa com 1mm obs')
print(Data_Jobs_Prob['Diff_1mm'].mean())
```
