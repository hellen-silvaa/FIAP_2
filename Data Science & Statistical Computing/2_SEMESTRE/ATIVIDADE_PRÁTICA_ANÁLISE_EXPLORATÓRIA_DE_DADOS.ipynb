```python
# bibliotecas para análise e gráficos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# abrir o dataframe
from bokeh.sampledata.iris import flowers as data
```

```python
# primeiras 5 linhas
display(data.head())

# dimensões (linhas, colunas)
print("Dimensões (linhas, colunas):", data.shape)

# tipos das colunas
print("\nTipos de dados:")
print(data.dtypes)

# nomes das colunas
print("\nNomes das colunas:", list(data.columns))
```

```python
# garantir estilo dos gráficos
sns.set(style="whitegrid")
```

```python
#2.  Exiba um resumo das principais estatísticas do banco de dados.
display(data.describe())
```

```python
#3. 3.Exiba a frequência absoluta das espécies.

freq_abs = data['species'].value_counts()
print("Frequência absoluta:\n", freq_abs)
```

```python
#4. Exiba a frequência relativa das espécies.

freq_rel = data['species'].value_counts(normalize=True)
print("Frequência relativa:\n", freq_rel)

```

```python
#5. Exiba um gráfico de barras (verticais), mostrando o total por espécies.
plt.figure(figsize=(6,4))
sns.countplot(x='species', data=data, order=data['species'].value_counts().index)
plt.title('Total por espécie (barras verticais)')
plt.xlabel('Espécie')
plt.ylabel('Contagem')
plt.tight_layout()
plt.show()
```

```python
#6. Exiba um gráfico de colunas (horizontais), mostrando o total por espécies.
plt.figure(figsize=(6,4))
order = data['species'].value_counts().index
sns.countplot(y='species', data=data, order=order)
plt.title('Total por espécie (barras horizontais)')
plt.xlabel('Contagem')
plt.ylabel('Espécie')
plt.tight_layout()
plt.show()
```

```python
#7. Exiba bloxplot do dataset (mostrando de todas as colunas).
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_melt = data.melt(id_vars='species', value_vars=numeric_cols,
                    var_name='medida', value_name='valor')

plt.figure(figsize=(10,6))
sns.boxplot(x='medida', y='valor', hue='species', data=df_melt)
plt.title('Boxplots por medida e espécie')
plt.legend(title='Espécie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#8.Calcule a média e mediana agrupadas por espécies.
agg_mean_median = data.groupby('species')[numeric_cols].agg(['mean','median'])
display(agg_mean_median)
```

```python
#9. Calcule o desvio padrão e a variância agrupadas por espécies.
agg_std_var = data.groupby('species')[numeric_cols].agg(['std','var'])
display(agg_std_var)
```
