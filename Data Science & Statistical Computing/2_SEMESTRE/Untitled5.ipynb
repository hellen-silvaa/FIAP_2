```python
# CLASSE OCORRÃŠNCIA
@dataclass
class Ocorrencia:
    """
    representa uma ocorrÃªncia de incÃªndio no nosso sistema Curupira Digital

    Attributes:
        id: Identificador Ãºnico da ocorrÃªncia
        descricao: DescriÃ§Ã£o do tipo de incÃªndio
        prioridade: NÃ­vel de prioridade (1-10, onde 10 Ã© crÃ­tico)
        localizacao: Coordenadas ou descriÃ§Ã£o do local
        timestamp: Momento da ocorrÃªncia
        status: Status atual (Pendente, Em Atendimento, Resolvido)
        equipe_responsavel: ID da equipe designada
    """
    id: int
    descricao: str
    prioridade: int
    localizacao: str
    timestamp: datetime
    status: str = "Pendente"
    equipe_responsavel: Optional[str] = None

    def __lt__(self, other):
        """ComparaÃ§Ã£o para heap (prioridade maior primeiro)"""
        return self.prioridade > other.prioridade

    def __str__(self):
        return (f" ID: {self.id}\n"
                f" DescriÃ§Ã£o: {self.descricao}\n "
                f" Prioridade: {self.prioridade}\n"
                f" LocalizaÃ§Ã£o: {self.localizacao}")

print("âœ… Classe Ocorrencia criada com sucesso!")
teste_ocorrencia = Ocorrencia(
    id=1,
    descricao="IncÃªndio Florestal",
    prioridade=8,
    localizacao="RegiÃ£o Norte",
    timestamp=datetime.now()
)
print(f"ğŸ“ Exemplo: {teste_ocorrencia}")
```

```python
#  HEAP DE PRIORIDADES

class HeapPrioridade:
    """
    ESTRUTURA 1: HEAP
    fila de prioridade para gerenciar emergÃªncias crÃ­ticas recebidas
    complexidade: O(log n) para inserÃ§Ã£o e remoÃ§Ã£o
    """

    def __init__(self):
        """inicializa heap vazio"""
        self.heap = []
        self.contador_insercoes = 0

    def inserir_emergencia(self, ocorrencia: Ocorrencia) -> None:
        """
        insere uma nova emergÃªncia mantendo a propriedade do heap
        Args:
            ocorrencia: objeto ocorrencia a ser inserido
        complexidade: O(log n)
        """
        heapq.heappush(self.heap, ocorrencia)
        self.contador_insercoes += 1
        print(f" EMERGÃŠNCIA INSERIDA: {ocorrencia.descricao} (Prioridade: {ocorrencia.prioridade})")

    def atender_emergencia_critica(self) -> Optional[Ocorrencia]:
        """
        nessa parte remove e retorna emergÃªncia de maior prioridade
        Returns:
            ocorrencia de maior prioridade ou none se estiver vazio
        Complexidade: O(log n)
        """
        if self.heap:
            emergencia = heapq.heappop(self.heap)
            emergencia.status = "Em Atendimento"
            print(f" ATENDENDO EMERGÃŠNCIA CRÃTICA: {emergencia.descricao}")
            return emergencia
        print("âš ï¸ Nenhuma emergÃªncia pendente!")
        return None

    def visualizar_heap(self) -> None:
        """distinado a exibir o estado atual do heap de prioridades"""
        print(f"\nğŸ”¥ === HEAP DE EMERGÃŠNCIAS ({len(self.heap)} itens) ===")
        if not self.heap:
            print("   ğŸ“­ Nenhuma emergÃªncia pendente")
            return

        # ordena por prioridade para visualizaÃ§Ã£o e sem alterar heap
        emergencias_ordenadas = sorted(self.heap, reverse=True)
        for i, ocorrencia in enumerate(emergencias_ordenadas, 1):
            print(f"   {i}.  Prioridade {ocorrencia.prioridade}: {ocorrencia.descricao}")

    def esta_vazio(self) -> bool:
        #verifica se heap estÃ¡ vazio
        return len(self.heap) == 0

    def tamanho(self) -> int:
        #retorna nÃºmero de elementos no heap
        return len(self.heap)


print("ğŸš¨ DEMONSTRAÃ‡ÃƒO: HEAP DE PRIORIDADES ğŸš¨")
heap_emergencias = HeapPrioridade()

# inserindo emergÃªncias com diferentes prioridades
emergencias_teste = [
    Ocorrencia(1, "IncÃªndio Urbano", 9, "Parque da aclimaÃ§Ã£o", datetime.now()),
    Ocorrencia(2, "Queimada Rural", 5, "Zona Rural", datetime.now()),
    Ocorrencia(3, "IncÃªndio Industrial", 8, "Distrito Industrial", datetime.now()),
    Ocorrencia(4, "Foco Pequeno", 3, "Bairro mococa interior de sÃ£o paulo", datetime.now()),
    Ocorrencia(5, "IncÃªndio Florestal CrÃ­tico", 10, "Reserva Florestal do Instituto de BiociÃªncias (RFIB)", datetime.now()),
]

for emergencia in emergencias_teste:
    heap_emergencias.inserir_emergencia(emergencia)

heap_emergencias.visualizar_heap()

print("\n ğŸƒâ€â™‚ï¸ Atendendo emergÃªncias por ordem de prioridade: ğŸƒâ€â™‚ï¸")
while not heap_emergencias.esta_vazio():
    emergencia = heap_emergencias.atender_emergencia_critica()
    time.sleep(0.5)  # Simula tempo de processamento
```

```python
#FILA DE ATENDIMENTO
from collections import deque

class FilaAtendimento:
    """
    ESTRUTURA 2: FILA (FIFO - First In, First Out)
    gerencia atendimento sequencial de ocorrÃªncias nÃ£o crÃ­ticas
    Complexidade: O(1) para inserÃ§Ã£o e remoÃ§Ã£o
    """

    def __init__(self, nome_fila: str):
        """
        inicializa fila de atendimento
        Args:
            nome_fila: nome identificador da fila
        """
        self.fila = deque()
        self.nome = nome_fila
        self.total_atendidos = 0

    def adicionar_ocorrencia(self, ocorrencia: Ocorrencia) -> None:
        """
        parte que adiciona ocorrÃªncia ao final da fila
        Args:
            ocorrencia: objeto ocorrencia a ser enfileirado
        Complexidade: O(1)
        """
        self.fila.append(ocorrencia)
        print(f"ğŸ“‹ ADICIONADO Ã€ FILA {self.nome}: {ocorrencia.descricao}")

    def atender_proximo(self) -> Optional[Ocorrencia]:
        """
        remove e retorna primeira ocorrÃªncia da fila
        Returns:
            primeira ocorrÃªncia ou none se fila vazia
        Complexidade: O(1)
        """
        if self.fila:
            ocorrencia = self.fila.popleft()
            ocorrencia.status = "Em Atendimento"
            self.total_atendidos += 1
            print(f"ğŸ‘¨â€ğŸš’ ATENDENDO DA FILA {self.nome}: {ocorrencia.descricao}")
            return ocorrencia
        print(f"ğŸ“­ Fila {self.nome} estÃ¡ vazia!")
        return None

    def visualizar_fila(self) -> None:
        #exibe estado atual da fila
        print(f"\nğŸ“‹ === FILA {self.nome} ({len(self.fila)} itens) ===")
        if not self.fila:
            print("   ğŸ“­ Fila vazia")
            return

        print("   ğŸ“ PrÃ³ximos atendimentos:")
        for i, ocorrencia in enumerate(self.fila, 1):
            marcador = "â¤" if i == 1 else " "
            print(f"   {marcador} {i}. {ocorrencia.descricao} (ID: {ocorrencia.id})")

    def tamanho(self) -> int:
        #retorna nÃºmero de elementos na fila
        return len(self.fila)

    def esta_vazia(self) -> bool:
        #verifica se estÃ¡ vazia a fila
        return len(self.fila) == 0

    def estatisticas(self) -> Dict:
        return {
            "nome": self.nome,
            "pendentes": len(self.fila),
            "atendidos": self.total_atendidos,
            "total_processados": len(self.fila) + self.total_atendidos
        }

print("ğŸ“‹ === DEMONSTRAÃ‡ÃƒO: FILA DE ATENDIMENTO ===")

fila_norte = FilaAtendimento("REGIÃƒO NORTE")
fila_sul = FilaAtendimento("REGIÃƒO SUL")

# simulando a inserÃ§Ã£o de ocorrÃªncias Ã s filas
ocorrencias_norte = [
    Ocorrencia(10, "Queimada Controlada", 4, "Fazenda Norte", datetime.now()),
    Ocorrencia(11, "Foco em Pasto", 6, "RegiÃ£o Norte-2", datetime.now()),
    Ocorrencia(12, "IncÃªndio Residencial", 7, "Bairro Norte", datetime.now())
]

ocorrencias_sul = [
    Ocorrencia(20, "Queimada AgrÃ­cola", 5, "PlantaÃ§Ã£o Sul", datetime.now()),
    Ocorrencia(21, "Foco em Eucalipto", 6, "RegiÃ£o Sul-1", datetime.now())
]

print("\nğŸŒ Distribuindo ocorrÃªncias por regiÃ£o:")
for ocorrencia in ocorrencias_norte:
    fila_norte.adicionar_ocorrencia(ocorrencia)

for ocorrencia in ocorrencias_sul:
    fila_sul.adicionar_ocorrencia(ocorrencia)

# visualizando filas
fila_norte.visualizar_fila()
fila_sul.visualizar_fila()

print("\nğŸ‘¨â€ğŸš’ Processando atendimentos sequenciais:")
for _ in range(2):
    fila_norte.atender_proximo()
    fila_sul.atender_proximo()
    time.sleep(0.3)

print(f"\nğŸ“Š EstatÃ­sticas Norte: {fila_norte.estatisticas()}")
print(f"ğŸ“Š EstatÃ­sticas Sul: {fila_sul.estatisticas()}")
```

```python
#PILHA DE HISTÃ“RICO
class PilhaHistorico:
    """
    ESTRUTURA 3: PILHA (LIFO - Last In, First Out)
    ggerencia histÃ³rico de aÃ§Ãµes realizadas pelas equipes
    Complexidade: O(1) para inserÃ§Ã£o e remoÃ§Ã£o
    """

    def __init__(self, equipe_id: str):
        """
        inicializa pilha de histÃ³rico para uma equipe
        Args:
            equipe_id: Identificador da equipe
        """
        self.pilha = []
        self.equipe_id = equipe_id
        self.contador_acoes = 0

    def registrar_acao(self, acao: str, detalhes: str = "") -> None:
        """
        registra uma nova aÃ§Ã£o no topo da pilha
        Args:
            acao: descriÃ§Ã£o da aÃ§Ã£o realizada
            detalhes: detalhes adicionais da aÃ§Ã£o
        Complexidade: O(1)
        """
        registro = {
            "timestamp": datetime.now(),
            "acao": acao,
            "detalhes": detalhes,
            "numero_acao": self.contador_acoes + 1
        }
        self.pilha.append(registro)
        self.contador_acoes += 1
        print(f"ğŸ“ AÃ‡ÃƒO REGISTRADA [{self.equipe_id}]: {acao}")

    def desfazer_ultima_acao(self) -> Optional[Dict]:
        """
        remove e retorna Ãºltima aÃ§Ã£o (simulando o desfazer)
        Returns:
            uÂ´ltimo registro ou none se pilha vazia
        Complexidade: O(1)
        """
        if self.pilha:
            acao_desfeita = self.pilha.pop()
            print(f"â†©ï¸ AÃ‡ÃƒO DESFEITA [{self.equipe_id}]: {acao_desfeita['acao']}")
            return acao_desfeita
        print(f"âš ï¸ Nenhuma aÃ§Ã£o para desfazer em {self.equipe_id}")
        return None

    def visualizar_historico(self, limite: int = 10) -> None:
        """
        exibe histÃ³rico de aÃ§Ãµes (do mais recente ao mais antigo)
        Args:
            limite: NÃºmero mÃ¡ximo de aÃ§Ãµes a exibir
        """
        print(f"\nğŸ“š === HISTÃ“RICO {self.equipe_id} ===")
        if not self.pilha:
            print("   ğŸ“­ Nenhuma aÃ§Ã£o registrada")
            return

        print(f"   ğŸ“Š Total de aÃ§Ãµes: {len(self.pilha)}")
        print("   ğŸ•’ Ãšltimas aÃ§Ãµes (mais recente primeiro):")

        # mostra as Ãºltimas aÃ§Ãµes  Ã© o topo da pilha primeiro
        for i, registro in enumerate(reversed(self.pilha[-limite:]), 1):
            tempo = registro["timestamp"].strftime("%H:%M:%S")
            print(f"   {i}. [{tempo}] {registro['acao']}")
            if registro["detalhes"]:
                print(f"      ğŸ’¬ {registro['detalhes']}")

    def buscar_acao(self, termo_busca: str) -> List[Dict]:
        """
        busca aÃ§Ãµes que contenham o termo especificado
        Args:
            termo_busca: Termo a ser buscado nas aÃ§Ãµes
        Returns:
            lista de registros que contÃªm o termo
        Complexidade: O(n)
        """
        resultados = []
        for registro in self.pilha:
            if termo_busca.lower() in registro["acao"].lower():
                resultados.append(registro)
        return resultados

    def estatisticas(self) -> Dict:
        #retorna estatÃ­sticas do histÃ³rico
        return {
            "equipe": self.equipe_id,
            "total_acoes": len(self.pilha),
            "primeira_acao": self.pilha[0]["timestamp"] if self.pilha else None,
            "ultima_acao": self.pilha[-1]["timestamp"] if self.pilha else None
        }

    def esta_vazia(self) -> bool:
        #verifica se pilha estÃ¡ vazia
        return len(self.pilha) == 0

print("ğŸ“š === DEMONSTRAÃ‡ÃƒO: PILHA DE HISTÃ“RICO ===")

# criando pilhas de histÃ³rico para diferentes equipes
historico_equipe_CurupiraDigital = PilhaHistorico("EQUIPE CurupiraDigital")
historico_equipe_Codebreakers = PilhaHistorico("EQUIPE Codebreakers")

#simulando o trabalho das equipes de bombeiros
print("\nğŸš’ Simulando aÃ§Ãµes da Equipe CurupiraDigital:")
acoes_CurupiraDigital = [
    ("Chegada ao local", "IncÃªndio florestal na Reserva Norte"),
    ("AvaliaÃ§Ã£o inicial", "Ãrea afetada: 2 hectares, vento moderado"),
    ("InÃ­cio do combate", "Utilizando abafadores e Ã¡gua"),
    ("SolicitaÃ§Ã£o de reforÃ§o", "Fogo se espalhando rapidamente"),
    ("ReforÃ§o chegou", "Equipe Codebreakers se juntou ao combate"),
    ("Fogo controlado", "PerÃ­metro estabelecido com sucesso")
]

for acao, detalhe in acoes_CurupiraDigital:
    historico_equipe_CurupiraDigital.registrar_acao(acao, detalhe)
    time.sleep(0.2)

print("\nğŸš’ Simulando aÃ§Ãµes da Equipe Codebreakers:")
acoes_Codebreakers = [
    ("Deslocamento", "Indo para apoiar Equipe CurupiraDigital"),
    ("Chegada ao reforÃ§o", "Posicionamento no flanco leste"),
    ("Combate coordenado", "Atacando fogo por duas frentes"),
    ("Rescaldo iniciado", "Eliminando focos residuais")
]

for acao, detalhe in acoes_Codebreakers:
    historico_equipe_Codebreakers.registrar_acao(acao, detalhe)
    time.sleep(0.2)

# visualizando os histÃ³ricos
historico_equipe_CurupiraDigital.visualizar_historico()
historico_equipe_Codebreakers.visualizar_historico()

# vemonstrando funcionalidade de desfazer
print("\nâ†©ï¸ Testando funcionalidade de desfazer:")
historico_equipe_CurupiraDigital.desfazer_ultima_acao()
historico_equipe_CurupiraDigital.visualizar_historico(3)

# busca no histÃ³rico
print("\nğŸ” Buscando aÃ§Ãµes relacionadas a 'combate':")
resultados = historico_equipe_CurupiraDigital.buscar_acao("combate")
for resultado in resultados:
    print(f"   ğŸ¯ {resultado['acao']}: {resultado['detalhes']}")
```

```python
# ÃRVORE DE REGIÃ•ES
class NoRegiao:
    """
    nÃ³ da Ã¡rvore representando uma regiÃ£o geogrÃ¡fica
    """
    def __init__(self, nome: str, nivel: int):
        """
        inicializa nÃ³ da regiÃ£o
        Args:
            nome: Nome da regiÃ£o
            nivel: nÃ­vel hierÃ¡rquico (0=paÃ­s, 1=estado, 2=cidade, etc.)
        """
        self.nome = nome
        self.nivel = nivel
        self.filhos = []
        self.ocorrencias = []
        self.equipes_alocadas = []
        self.estatisticas = {"total_ocorrencias": 0, "ocorrencias_resolvidas": 0}

    def adicionar_filho(self, filho):
        #adiciona regiÃ£o filha
        self.filhos.append(filho)

    def adicionar_ocorrencia(self, ocorrencia: Ocorrencia):
        #adiciona ocorrÃªncia Ã  regiÃ£o
        self.ocorrencias.append(ocorrencia)
        self.estatisticas["total_ocorrencias"] += 1

    def resolver_ocorrencia(self, ocorrencia_id: int):
        #marca ocorrÃªncia como resolvida
        for ocorrencia in self.ocorrencias:
            if ocorrencia.id == ocorrencia_id:
                ocorrencia.status = "Resolvido"
                self.estatisticas["ocorrencias_resolvidas"] += 1
                break

class ArvoreRegioes:
    """
    ESTRUTURA 4: ÃRVORE
    hierarquia de regiÃµes para organizaÃ§Ã£o geogrÃ¡fica
    permite navegaÃ§Ã£o e busca hierÃ¡rquica eficiente
    """

    def __init__(self):
        #inicializa Ã¡rvore de regiÃµes
        self.raiz = NoRegiao("BRASIL", 0)
        self.total_nos = 1

    def adicionar_regiao(self, caminho: List[str]) -> NoRegiao:
        """
        adiciona regiÃ£o seguindo caminho hierÃ¡rquico
        Args:
            caminho: lista com hierarquia [paÃ­s, estado, cidade, bairro]
        Returns:
            nÃ³ da regiÃ£o criada/encontrada
        Complexidade: O(d) onde d Ã© a profundidade
        """
        no_atual = self.raiz

        for i, nome_regiao in enumerate(caminho[1:], 1):  # deve pular o paÃ­s (raiz)
            # busca se a regiÃ£o jÃ¡ existe nos filhos
            no_encontrado = None
            for filho in no_atual.filhos:
                if filho.nome == nome_regiao:
                    no_encontrado = filho
                    break

            # se nÃ£o existir, cria um novo nÃ³
            if not no_encontrado:
                novo_no = NoRegiao(nome_regiao, i)
                no_atual.adicionar_filho(novo_no)
                no_atual = novo_no
                self.total_nos += 1
                print(f"ğŸŒ REGIÃƒO CRIADA: {' > '.join(caminho[:i+1])}")
            else:
                no_atual = no_encontrado

        return no_atual

    def buscar_regiao(self, nome: str) -> List[NoRegiao]:
        """
        busca regiÃµes por nome (busca em profundidade)
        Args:
            nome: bome da regiÃ£o a buscar
        Returns:
            lista de nÃ³s encontrados
        Complexidade: O(n) onde n Ã© nÃºmero de nÃ³s
        """
        resultados = []

        def dfs(no_atual):
            if nome.lower() in no_atual.nome.lower():
                resultados.append(no_atual)
            for filho in no_atual.filhos:
                dfs(filho)

        dfs(self.raiz)
        return resultados

    def visualizar_arvore(self, no_atual=None, profundidade=0) -> None:
        """
        visualiza Ã¡rvore hierÃ¡rquica
        Args:
            no_atual: nÃ³ atual (padrÃ£o: raiz)
            profundidade: profundidade atual na recursÃ£o
        """
        if no_atual is None:
            no_atual = self.raiz
            print(f"\nğŸŒ³ === ÃRVORE DE REGIÃ•ES ({self.total_nos} nÃ³s) ===")

        # indentaÃ§Ã£o baseada na profundidade
        indent = "  " * profundidade
        marcador = "ğŸŒ" if profundidade == 0 else "ğŸ“"

        # informaÃ§Ãµes do nÃ³
        total_ocorr = no_atual.estatisticas["total_ocorrencias"]
        resolvidas = no_atual.estatisticas["ocorrencias_resolvidas"]

        print(f"{indent}{marcador} {no_atual.nome} (OcorrÃªncias: {total_ocorr}, Resolvidas: {resolvidas})")

        # recursÃ£o para filhos
        for filho in no_atual.filhos:
            self.visualizar_arvore(filho, profundidade + 1)

    def gerar_relatorio_regiao(self, nome_regiao: str) -> Dict:
        """
        gera relatÃ³rio detalhado de uma regiÃ£o
        Args:
            nome_regiao: nome da regiÃ£o
        Returns:
            dicionÃ¡rio com estatÃ­sticas da regiÃ£o
        """
        regioes_encontradas = self.buscar_regiao(nome_regiao)
        if not regioes_encontradas:
            return {"erro": f"RegiÃ£o '{nome_regiao}' nÃ£o encontrada"}

        regiao = regioes_encontradas[0]  # primeira regiÃ£o encontrada

        # calcula estatÃ­sticas recursivamente -- incluindo as sub-regiÃµes
        def calcular_stats_recursivo(no):
            stats = {
                "total_ocorrencias": no.estatisticas["total_ocorrencias"],
                "ocorrencias_resolvidas": no.estatisticas["ocorrencias_resolvidas"],
                "sub_regioes": len(no.filhos)
            }

            # aqui soma estatÃ­sticas dos filhos
            for filho in no.filhos:
                stats_filho = calcular_stats_recursivo(filho)
                stats["total_ocorrencias"] += stats_filho["total_ocorrencias"]
                stats["ocorrencias_resolvidas"] += stats_filho["ocorrencias_resolvidas"]

            return stats

        return {
            "regiao": regiao.nome,
            "nivel": regiao.nivel,
            "estatisticas": calcular_stats_recursivo(regiao)
        }

    def listar_caminho_para_regiao(self, nome_regiao: str) -> List[str]:
        """
        encontra caminho da raiz atÃ© a regiÃ£o especificada
        Args:
            nome_regiao: nome da regiÃ£o de destino
        Returns:
            lista com caminho hierÃ¡rquico
        """
        def dfs_caminho(no_atual, caminho_atual):
            caminho_atual.append(no_atual.nome)

            if no_atual.nome.lower() == nome_regiao.lower():
                return caminho_atual.copy()

            for filho in no_atual.filhos:
                resultado = dfs_caminho(filho, caminho_atual)
                if resultado:
                    return resultado

            caminho_atual.pop()
            return None

        return dfs_caminho(self.raiz, [])

# demonstraÃ§Ã£o da Ãrvore
print("ğŸŒ³ === DEMONSTRAÃ‡ÃƒO: ÃRVORE DE REGIÃ•ES ===")

# criando Ã¡rvore de regiÃµes do Brasil
arvore_brasil = ArvoreRegioes()

# adicionando regiÃµes baseado em dados do INPE do histÃ³rico de focos de incendio
regioes_brasil = [
    ["Brasil", "Mato Grosso", "CuiabÃ¡", "VÃ¡rzea Grande"],
    ["Brasil", "ParÃ¡", "Altamira", "TransAmazÃ´nica"],
    ["Brasil", "Amazonas", "LÃ¡brea", "Sul do Amazonas"],
    ["Brasil", "Tocantins", "AraguaÃ­na", "Norte Tocantinense"],
    ["Brasil", "MaranhÃ£o", "Imperatriz", "Oeste Maranhense"],
    ["Brasil", "RondÃ´nia", "Porto Velho", "Madeira-MamorÃ©"],
    ["Brasil", "Mato Grosso do Sul", "CorumbÃ¡", "Pantanal Sul"],
    ["Brasil", "GoiÃ¡s", "Rio Verde", "Sudoeste Goiano"],
]

print("\nğŸ—ï¸ Construindo hierarquia de regiÃµes:")
nos_regioes = {}
for caminho in regioes_brasil:
    no_regiao = arvore_brasil.adicionar_regiao(caminho)
    nos_regioes[caminho[-1]] = no_regiao  # guarda a referÃªncia da regiÃ£o final

# Adicionando ocorrÃªncias Ã s regiÃµes crÃ­ticas baseadas no INPE
print("\nğŸ”¥ Distribuindo ocorrÃªncias por regiÃµes crÃ­ticas do Brasil:")
print("ğŸ“Š Baseado em padrÃµes reais de focos de calor por bioma\n")

ocorrencias_por_regiao = [
    # MT - 40% dos focos nacionais (Cerrado + Pantanal + AmazÃ´nia)
    ("VÃ¡rzea Grande", Ocorrencia(100, "IncÃªndio Florestal - Cerrado", 9,
                                "VÃ¡rzea Grande/MT - Coord: -15.6467,-56.1326", datetime.now())),

    # PA - 15% dos focos (AmazÃ´nia + Desmatamento)
    ("TransAmazÃ´nica", Ocorrencia(101, "Queimada Ilegal - Floresta PrimÃ¡ria", 10,
                                 "TransAmazÃ´nica/PA - Coord: -8.7619,-54.9306", datetime.now())),

    # AM - 10% dos focos (AmazÃ´nia Central)
    ("Sul do Amazonas", Ocorrencia(102, "IncÃªndio em UC - FLONA", 10,
                                  "Sul do Amazonas/AM - Coord: -8.1689,-63.0136", datetime.now())),

    # TO - 6% dos focos (Cerrado de TransiÃ§Ã£o)
    ("Norte Tocantinense", Ocorrencia(103, "Queimada no Cerrado - Savana", 8,
                                     "Norte Tocantinense/TO - Coord: -7.1195,-48.2072", datetime.now())),

    # MA - 5% dos focos (Cerrado Maranhense)
    ("Oeste Maranhense", Ocorrencia(104, "Foco em FormaÃ§Ã£o SavÃ¢nica", 7,
                                   "Oeste Maranhense/MA - Coord: -5.5242,-47.4804", datetime.now())),

    # RO - 4% dos focos (AmazÃ´nia Legal)
    ("Madeira-MamorÃ©", Ocorrencia(105, "IncÃªndio Rural - Pastagem", 8,
                                 "Madeira-MamorÃ©/RO - Coord: -8.7619,-63.9004", datetime.now())),

    # MS - 3% dos focos (Pantanal + Cerrado)
    ("Pantanal Sul", Ocorrencia(106, "IncÃªndio Pantanal - Wetland", 10,
                               "Pantanal Sul/MS - Coord: -19.0208,-57.6533", datetime.now())),

    # GO - 3% dos focos (Cerrado Central)
    ("Sudoeste Goiano", Ocorrencia(107, "Queimada AgrÃ­cola - PÃ³s-Colheita", 6,
                                  "Sudoeste Goiano/GO - Coord: -17.8011,-50.9208", datetime.now()))
]

for nome_regiao, ocorrencia in ocorrencias_por_regiao:
    if nome_regiao in nos_regioes:
        nos_regioes[nome_regiao].adicionar_ocorrencia(ocorrencia)
        print(f"ğŸ“ OcorrÃªncia adicionada Ã  {nome_regiao}: {ocorrencia.descricao}")

# Resolvendo algumas ocorrÃªncias (CORRIGIDO - usando IDs corretos!)
print("\nâœ… Resolvendo algumas ocorrÃªncias:")
nos_regioes["Sul do Amazonas"].resolver_ocorrencia(102)  # ID 102 estÃ¡ no Sul do Amazonas âœ…
nos_regioes["Sudoeste Goiano"].resolver_ocorrencia(107)  # ID 107 estÃ¡ no Sudoeste Goiano âœ…
nos_regioes["Pantanal Sul"].resolver_ocorrencia(106)     # ID 106 estÃ¡ no Pantanal Sul âœ…

# Visualizando Ã¡rvore completa
arvore_brasil.visualizar_arvore()

# Testando busca (CORRIGIDO - usando termos que existem!)
print("\nğŸ” Buscando regiÃµes com 'Mato':")
resultados_busca = arvore_brasil.buscar_regiao("Mato")
for resultado in resultados_busca:
    print(f"   ğŸ¯ Encontrado: {resultado.nome} (NÃ­vel {resultado.nivel})")

print("\nğŸ” Buscando regiÃµes com 'Amazonas':")
resultados_busca_am = arvore_brasil.buscar_regiao("Amazonas")
for resultado in resultados_busca_am:
    print(f"   ğŸ¯ Encontrado: {resultado.nome} (NÃ­vel {resultado.nivel})")

# Gerando relatÃ³rio de regiÃ£o (CORRIGIDO - usando estados que existem!)
print("\nğŸ“Š RelatÃ³rio detalhado do Mato Grosso:")
relatorio_mt = arvore_brasil.gerar_relatorio_regiao("Mato Grosso")
print(f"   ğŸ“ˆ {relatorio_mt}")

print("\nğŸ“Š RelatÃ³rio detalhado do ParÃ¡:")
relatorio_pa = arvore_brasil.gerar_relatorio_regiao("ParÃ¡")
print(f"   ğŸ“ˆ {relatorio_pa}")

# Testando caminho para regiÃ£o (CORRIGIDO - usando regiÃµes que existem!)
print("\nğŸ—ºï¸ Caminho para VÃ¡rzea Grande:")
caminho_vg = arvore_brasil.listar_caminho_para_regiao("VÃ¡rzea Grande")
if caminho_vg:
    print(f"   ğŸ“ Caminho: {' â†’ '.join(caminho_vg)}")
else:
    print("   âŒ Caminho nÃ£o encontrado")

print("\nğŸ—ºï¸ Caminho para TransAmazÃ´nica:")
caminho_ta = arvore_brasil.listar_caminho_para_regiao("TransAmazÃ´nica")
if caminho_ta:
    print(f"   ğŸ“ Caminho: {' â†’ '.join(caminho_ta)}")
else:
    print("   âŒ Caminho nÃ£o encontrado")

print("\nğŸ—ºï¸ Caminho para Pantanal Sul:")
caminho_ps = arvore_brasil.listar_caminho_para_regiao("Pantanal Sul")
if caminho_ps:
    print(f"   ğŸ“ Caminho: {' â†’ '.join(caminho_ps)}")
else:
    print("   âŒ Caminho nÃ£o encontrado")

# DemonstraÃ§Ã£o adicional - Listando todas as regiÃµes cadastradas
print("\nğŸŒ === TODAS AS REGIÃ•ES CADASTRADAS ===")
for nome_regiao in nos_regioes.keys():
    caminho = arvore_brasil.listar_caminho_para_regiao(nome_regiao)
    if caminho:
        print(f"   ğŸ“ {' â†’ '.join(caminho)}")

# EstatÃ­sticas finais por regiÃ£o crÃ­tica
print("\nğŸ“Š === ESTATÃSTICAS DAS REGIÃ•ES CRÃTICAS ===")
for nome_regiao in nos_regioes.keys():
    relatorio = arvore_brasil.gerar_relatorio_regiao(nome_regiao)
    if "erro" not in relatorio:
        stats = relatorio["estatisticas"]
        print(f"ğŸ”¥ {nome_regiao}: {stats['total_ocorrencias']} ocorrÃªncias, {stats['ocorrencias_resolvidas']} resolvidas")
```

```python
# LISTA LIGADA DE EQUIPES - VERSÃƒO LIMPA
class NoEquipe:
    """
    nÃ³ da lista ligada representando uma equipe de combate
    """
    def __init__(self, equipe_id: str, nome: str, localizacao: str, especialidade: str):
        """
        inicializa nÃ³ da equipe
        Args:
            equipe_id: Identificador Ãºnico da equipe
            nome: Nome da equipe
            localizacao: LocalizaÃ§Ã£o atual da equipe
            especialidade: Especialidade da equipe (florestal, urbano, etc.)
        """
        self.equipe_id = equipe_id
        self.nome = nome
        self.localizacao = localizacao
        self.especialidade = especialidade
        self.status = "DisponÃ­vel"
        self.ocorrencia_atual = None
        self.proximo = None  # o ponteiro para prÃ³xima equipe
        self.historico_atendimentos = []

    def __str__(self):
        status_icon = "ğŸŸ¢" if self.status == "DisponÃ­vel" else "ğŸ”´"
        return f"{status_icon} {self.nome} ({self.especialidade}) - {self.localizacao}"

class ListaLigadaEquipes:
    """
    ESTRUTURA 5: LISTA LIGADA
    gerencia equipes conectadas por proximidade geogrÃ¡fica
    permite navegaÃ§Ã£o sequencial e busca eficiente
    """

    def __init__(self):
        self.cabeca = None
        self.tamanho = 0

    def adicionar_equipe(self, equipe_id: str, nome: str, localizacao: str, especialidade: str) -> NoEquipe:
        """
        adiciona nova equipe no inÃ­cio da lista
        Args:
            equipe_id: ID Ãºnico da equipe
            nome: Nome da equipe
            localizacao: LocalizaÃ§Ã£o da equipe
            especialidade: Especialidade da equipe
        Returns:
            nÃ³ da equipe criada
        complexidade: O(1)
        """
        nova_equipe = NoEquipe(equipe_id, nome, localizacao, especialidade)

        # insere no inÃ­cio da lista
        nova_equipe.proximo = self.cabeca
        self.cabeca = nova_equipe
        self.tamanho += 1

        return nova_equipe

    def buscar_equipe_por_id(self, equipe_id: str) -> Optional[NoEquipe]:
        """
        busca equipe por ID
        Args:
            equipe_id: ID da equipe a buscar
        Returns:
            nÃ³ da equipe ou None se nÃ£o encontrada
        complexidade: O(n)
        """
        atual = self.cabeca
        while atual:
            if atual.equipe_id == equipe_id:
                return atual
            atual = atual.proximo
        return None

    def buscar_equipes_disponiveis(self) -> List[NoEquipe]:
        """
        busca todas as equipes disponÃ­veis
        Returns:
            lista de equipes com status 'DisponÃ­vel'
        complexidade: O(n)
        """
        equipes_disponiveis = []
        atual = self.cabeca

        while atual:
            if atual.status == "DisponÃ­vel":
                equipes_disponiveis.append(atual)
            atual = atual.proximo

        return equipes_disponiveis

    def buscar_equipes_por_especialidade(self, especialidade: str) -> List[NoEquipe]:
        """
        busca equipes por especialidade
        Args:
            especialidade: Especialidade desejada
        Returns:
            lista de equipes com a especialidade
        complexidade: O(n)
        """
        equipes_especialidade = []
        atual = self.cabeca

        while atual:
            if atual.especialidade.lower() == especialidade.lower():
                equipes_especialidade.append(atual)
            atual = atual.proximo

        return equipes_especialidade

    def atribuir_ocorrencia(self, equipe_id: str, ocorrencia: Ocorrencia) -> bool:
        """
        atribui ocorrÃªncia a uma equipe especÃ­fica
        Args:
            equipe_id: ID da equipe
            ocorrencia: OcorrÃªncia a ser atribuÃ­da
        Returns:
            true se atribuiÃ§Ã£o foi bem-sucedida
        """
        equipe = self.buscar_equipe_por_id(equipe_id)
        if equipe and equipe.status == "DisponÃ­vel":
            equipe.status = "Em MissÃ£o"
            equipe.ocorrencia_atual = ocorrencia
            ocorrencia.equipe_responsavel = equipe_id
            ocorrencia.status = "Em Atendimento"
            return True
        return False

    def finalizar_atendimento(self, equipe_id: str) -> bool:
        """
        finaliza o atendimento de uma equipe
        Args:
            equipe_id: ID da equipe
        Returns:
            true se finalizaÃ§Ã£o foi bem-sucedida
        """
        equipe = self.buscar_equipe_por_id(equipe_id)
        if equipe and equipe.status == "Em MissÃ£o":
            # registra no histÃ³rico
            if equipe.ocorrencia_atual:
                equipe.historico_atendimentos.append({
                    "ocorrencia": equipe.ocorrencia_atual,
                    "timestamp_fim": datetime.now()
                })
                equipe.ocorrencia_atual.status = "Resolvido"

            equipe.status = "DisponÃ­vel"
            equipe.ocorrencia_atual = None
            return True
        return False

    def visualizar_lista(self) -> None:
        #Exibe todas as equipes da lista
        print(f"\nğŸš’ LISTA DE EQUIPES ({self.tamanho} equipes)")
        if not self.cabeca:
            print("  Nenhuma equipe cadastrada")
            return

        atual = self.cabeca
        posicao = 1

        while atual:
            print(f"  {posicao}. {atual}")
            if atual.ocorrencia_atual:
                print(f"     âš¡ {atual.ocorrencia_atual.descricao}")
            elif atual.historico_atendimentos:
                print(f"     ğŸ“Š {len(atual.historico_atendimentos)} atendimentos realizados")

            atual = atual.proximo
            posicao += 1

    def obter_estatisticas(self) -> Dict:
        """
        calcula estatÃ­sticas gerais das equipes
        Returns:
            dicionÃ¡rio com estatÃ­sticas
        """
        if not self.cabeca:
            return {"total": 0}

        stats = {
            "total": self.tamanho,
            "disponiveis": 0,
            "em_missao": 0,
            "por_especialidade": {},
            "total_atendimentos": 0
        }

        atual = self.cabeca
        while atual:
            #status
            if atual.status == "DisponÃ­vel":
                stats["disponiveis"] += 1
            else:
                stats["em_missao"] += 1

            #especialidade
            esp = atual.especialidade
            stats["por_especialidade"][esp] = stats["por_especialidade"].get(esp, 0) + 1

            #atendimentos
            stats["total_atendimentos"] += len(atual.historico_atendimentos)

            atual = atual.proximo

        return stats

    def remover_equipe(self, equipe_id: str) -> bool:
        """
        remove equipe da lista (apenas se disponÃ­vel)
        Args:
            equipe_id: ID da equipe a remover
        Returns:
            true se remoÃ§Ã£o foi bem-sucedida
        complexidade: O(n)
        """
        if not self.cabeca:
            return False

        #se Ã© o primeiro nÃ³
        if self.cabeca.equipe_id == equipe_id:
            if self.cabeca.status == "DisponÃ­vel":
                self.cabeca = self.cabeca.proximo
                self.tamanho -= 1
                return True
            return False

        #busca nos demais nÃ³s
        atual = self.cabeca
        while atual.proximo:
            if atual.proximo.equipe_id == equipe_id:
                if atual.proximo.status == "DisponÃ­vel":
                    atual.proximo = atual.proximo.proximo
                    self.tamanho -= 1
                    return True
                return False
            atual = atual.proximo

        return False

print("ğŸš’ DEMONSTRAÃ‡ÃƒO: LISTA LIGADA DE EQUIPES")

lista_equipes = ListaLigadaEquipes()
equipes_cadastro = [
    ("EQ001", "Bombeiros MT", "VÃ¡rzea Grande/MT", "Florestal"),
    ("EQ002", "Resgate PA", "Altamira/PA", "Florestal"),
    ("EQ003", "Florestal AM", "LÃ¡brea/AM", "Florestal"),
    ("EQ004", "Cerrado TO", "AraguaÃ­na/TO", "Rural"),
    ("EQ005", "ForÃ§a MA", "Imperatriz/MA", "Rural"),
    ("EQ006", "AmazÃ´nia RO", "Porto Velho/RO", "Florestal"),
    ("EQ007", "Pantanal MS", "CorumbÃ¡/MS", "Pantanal"),
    ("EQ008", "Agro GO", "Rio Verde/GO", "Rural")
]

print("\nCadastrando equipes especializadas...")
for equipe_id, nome, localizacao, especialidade in equipes_cadastro:
    lista_equipes.adicionar_equipe(equipe_id, nome, localizacao, especialidade)

#visualizaÃ§Ã£o inicial
lista_equipes.visualizar_lista()

#criando ocorrÃªncias para poder atribuir
ocorrencias_atribuicao = [
    Ocorrencia(200, "IncÃªndio Florestal CrÃ­tico", 9, "Mata AtlÃ¢ntica", datetime.now()),
    Ocorrencia(201, "Queimada no Cerrado", 8, "Savana", datetime.now()),
    Ocorrencia(202, "IncÃªndio no Pantanal", 10, "Wetland", datetime.now())
]

print("\nAtribuindo ocorrÃªncias por especialidade...")

#busca e atribui por especialidade
especializacoes = ["Florestal", "Rural", "Pantanal"]
for i, especialidade in enumerate(especializacoes):
    if i < len(ocorrencias_atribuicao):
        equipes_esp = lista_equipes.buscar_equipes_por_especialidade(especialidade)
        if equipes_esp:
            sucesso = lista_equipes.atribuir_ocorrencia(equipes_esp[0].equipe_id, ocorrencias_atribuicao[i])
            if sucesso:
                print(f"  {equipes_esp[0].nome} â†’ {ocorrencias_atribuicao[i].descricao}")

#estado apÃ³s atribuiÃ§Ãµes
lista_equipes.visualizar_lista()

#equipes disponÃ­veis
print("\nEquipes ainda disponÃ­veis:")
disponiveis = lista_equipes.buscar_equipes_disponiveis()
for equipe in disponiveis:
    print(f"  {equipe.nome} ({equipe.especialidade})")

#finalizando alguns atendimentos
print("\nFinalizando atendimentos...")
finalizadas = ["EQ001", "EQ004", "EQ007"]
for eq_id in finalizadas:
    if lista_equipes.finalizar_atendimento(eq_id):
        equipe = lista_equipes.buscar_equipe_por_id(eq_id)
        print(f"  {equipe.nome} finalizado")

#estatÃ­sticas finais resumidas
print("\nEstatÃ­sticas das equipes:")
stats = lista_equipes.obter_estatisticas()
print(f"  Total: {stats['total']} | DisponÃ­veis: {stats['disponiveis']} | Em missÃ£o: {stats['em_missao']}")
print(f"  Atendimentos realizados: {stats['total_atendimentos']}")
print(f"  Por especialidade: {stats['por_especialidade']}")
```

```python
#SISTEMA PRINCIPAL INTEGRADO
class CurupiraSimulator:
    """
    SISTEMA PRINCIPAL - CURUPIRA SIMULATOR
    Integra todas as 5 estruturas de dados para simular
    resposta coordenada a emergÃªncias de queimadas
    """

    def __init__(self):
        """Inicializa sistema completo"""
        print("ğŸŒ³ === INICIALIZANDO CURUPIRA SIMULATOR === ğŸŒ³")

        # Inicializando todas as estruturas
        self.heap_emergencias = HeapPrioridade()
        self.fila_norte = FilaAtendimento("NORTE")
        self.fila_sul = FilaAtendimento("SUL")
        self.fila_leste = FilaAtendimento("LESTE")
        self.fila_oeste = FilaAtendimento("OESTE")

        self.arvore_regioes = ArvoreRegioes()
        self.lista_equipes = ListaLigadaEquipes()

        # Pilhas de histÃ³rico por equipe (criadas dinamicamente)
        self.historicos = {}

        # Contadores gerais
        self.contador_ocorrencias = 0
        self.estatisticas_gerais = {
            "total_ocorrencias": 0,
            "emergencias_criticas": 0,
            "ocorrencias_resolvidas": 0,
            "tempo_resposta_medio": 0
        }

        print("âœ… Sistema inicializado com todas as estruturas!")
        self._configurar_sistema_inicial()

    def _configurar_sistema_inicial(self):
        """Configura estruturas iniciais do sistema"""
        print("\nğŸ—ï¸ Configurando estruturas iniciais...")

        # Configurando Ã¡rvore de regiÃµes
        regioes_iniciais = [
            ["Brasil", "SÃ£o Paulo", "Campinas"],
            ["Brasil", "SÃ£o Paulo", "SÃ£o Paulo"],
            ["Brasil", "Minas Gerais", "Belo Horizonte"],
            ["Brasil", "Rio de Janeiro", "Rio de Janeiro"],
            ["Brasil", "Mato Grosso", "CuiabÃ¡"]
        ]

        for caminho in regioes_iniciais:
            self.arvore_regioes.adicionar_regiao(caminho)

        # Configurando equipes iniciais
        equipes_iniciais = [
            ("ALPHA", "Bombeiros Alpha", "Norte", "Florestal"),
            ("BETA", "Bombeiros Beta", "Sul", "Urbano"),
            ("GAMMA", "Resgate Gamma", "Leste", "Industrial"),
            ("DELTA", "ForÃ§a Delta", "Oeste", "Florestal"),
            ("ECHO", "Equipe Echo", "Centro", "Rural")
        ]

        for eq_id, nome, loc, esp in equipes_iniciais:
            self.lista_equipes.adicionar_equipe(eq_id, nome, loc, esp)
            # Cria pilha de histÃ³rico para cada equipe
            self.historicos[eq_id] = PilhaHistorico(eq_id)

        print("âœ… ConfiguraÃ§Ã£o inicial concluÃ­da!")

    def criar_ocorrencia(self, descricao: str, prioridade: int, localizacao: str, regiao: str = "Norte") -> Ocorrencia:
        """
        Cria nova ocorrÃªncia e distribui pelas estruturas apropriadas
        Args:
            descricao: DescriÃ§Ã£o da ocorrÃªncia
            prioridade: NÃ­vel de prioridade (1-10)
            localizacao: Local da ocorrÃªncia
            regiao: RegiÃ£o geogrÃ¡fica
        Returns:
            Objeto Ocorrencia criado
        """
        self.contador_ocorrencias += 1
        ocorrencia = Ocorrencia(
            id=self.contador_ocorrencias,
            descricao=descricao,
            prioridade=prioridade,
            localizacao=localizacao,
            timestamp=datetime.now()
        )

        self.estatisticas_gerais["total_ocorrencias"] += 1

        # Se Ã© emergÃªncia crÃ­tica (prioridade >= 8), vai para heap
        if prioridade >= 8:
            self.heap_emergencias.inserir_emergencia(ocorrencia)
            self.estatisticas_gerais["emergencias_criticas"] += 1
        else:
            # Caso contrÃ¡rio, vai para fila regional
            filas_regionais = {
                "norte": self.fila_norte,
                "sul": self.fila_sul,
                "leste": self.fila_leste,
                "oeste": self.fila_oeste
            }

            fila = filas_regionais.get(regiao.lower(), self.fila_norte)
            fila.adicionar_ocorrencia(ocorrencia)

        print(f"ğŸ”¥ NOVA OCORRÃŠNCIA CRIADA: {ocorrencia}")
        return ocorrencia

    def processar_emergencias_criticas(self) -> None:
        """Processa todas as emergÃªncias crÃ­ticas pendentes"""
        print("\nğŸš¨ === PROCESSANDO EMERGÃŠNCIAS CRÃTICAS ===")

        emergencias_processadas = 0
        while not self.heap_emergencias.esta_vazio():
            emergencia = self.heap_emergencias.atender_emergencia_critica()

            # Busca melhor equipe para a emergÃªncia
            equipe_atribuida = self._atribuir_melhor_equipe(emergencia)

            if equipe_atribuida:
                # Registra aÃ§Ã£o no histÃ³rico da equipe
                self.historicos[equipe_atribuida.equipe_id].registrar_acao(
                    f"Atendimento emergÃªncia crÃ­tica",
                    f"{emergencia.descricao} em {emergencia.localizacao}"
                )
                emergencias_processadas += 1

            time.sleep(0.5)  # Simula tempo de processamento

        print(f"âœ… {emergencias_processadas} emergÃªncias crÃ­ticas processadas!")

    def processar_fila_regional(self, regiao: str) -> None:
        """
        Processa ocorrÃªncias de uma fila regional especÃ­fica
        Args:
            regiao: Nome da regiÃ£o (norte, sul, leste, oeste)
        """
        filas = {
            "norte": self.fila_norte,
            "sul": self.fila_sul,
            "leste": self.fila_leste,
            "oeste": self.fila_oeste
        }

        fila = filas.get(regiao.lower())
        if not fila:
            print(f"âŒ RegiÃ£o '{regiao}' nÃ£o encontrada")
            return

        print(f"\nğŸ“‹ === PROCESSANDO FILA {regiao.upper()} ===")

        processadas = 0
        while not fila.esta_vazia():
            ocorrencia = fila.atender_proximo()

            # Busca equipe disponÃ­vel na regiÃ£o
            equipes_regiao = self._buscar_equipes_por_regiao(regiao)
            equipe_disponivel = None

            for equipe in equipes_regiao:
                if equipe.status == "DisponÃ­vel":
                    equipe_disponivel = equipe
                    break

            if equipe_disponivel:
                self.lista_equipes.atribuir_ocorrencia(equipe_disponivel.equipe_id, ocorrencia)
                self.historicos[equipe_disponivel.equipe_id].registrar_acao(
                    f"Atendimento fila regional",
                    f"{ocorrencia.descricao} em {ocorrencia.localizacao}"
                )
                processadas += 1

                # Simula tempo de atendimento e finaliza
                time.sleep(0.3)
                self.lista_equipes.finalizar_atendimento(equipe_disponivel.equipe_id)
                self.estatisticas_gerais["ocorrencias_resolvidas"] += 1

        print(f"âœ… {processadas} ocorrÃªncias da regiÃ£o {regiao} processadas!")

    def _atribuir_melhor_equipe(self, ocorrencia: Ocorrencia) -> Optional[NoEquipe]:
        """
        Encontra e atribui a melhor equipe para uma ocorrÃªncia
        Args:
            ocorrencia: OcorrÃªncia a ser atribuÃ­da
        Returns:
            Equipe atribuÃ­da ou None se nenhuma disponÃ­vel
        """
        # Determina especialidade necessÃ¡ria baseada na descriÃ§Ã£o
        especialidade_necessaria = "Florestal"  # PadrÃ£o

        if "urbano" in ocorrencia.descricao.lower() or "prÃ©dio" in ocorrencia.descricao.lower():
            especialidade_necessaria = "Urbano"
        elif "industrial" in ocorrencia.descricao.lower() or "fÃ¡brica" in ocorrencia.descricao.lower():
            especialidade_necessaria = "Industrial"
        elif "rural" in ocorrencia.descricao.lower() or "fazenda" in ocorrencia.descricao.lower():
            especialidade_necessaria = "Rural"

        # Busca equipes da especialidade
        equipes_especialidade = self.lista_equipes.buscar_equipes_por_especialidade(especialidade_necessaria)

        # Se nÃ£o houver da especialidade, busca qualquer disponÃ­vel
        if not equipes_especialidade:
            equipes_especialidade = self.lista_equipes.buscar_equipes_disponiveis()

        # Atribui Ã  primeira disponÃ­vel
        for equipe in equipes_especialidade:
            if equipe.status == "DisponÃ­vel":
                self.lista_equipes.atribuir_ocorrencia(equipe.equipe_id, ocorrencia)
                return equipe

        print(f"âš ï¸ Nenhuma equipe disponÃ­vel para: {ocorrencia.descricao}")
        return None

    def _buscar_equipes_por_regiao(self, regiao: str) -> List[NoEquipe]:
        """
        Busca equipes por regiÃ£o (simulaÃ§Ã£o baseada na localizaÃ§Ã£o)
        Args:
            regiao: Nome da regiÃ£o
        Returns:
            Lista de equipes da regiÃ£o
        """
        equipes_regiao = []
        atual = self.lista_equipes.cabeca

        while atual:
            if regiao.lower() in atual.localizacao.lower():
                equipes_regiao.append(atual)
            atual = atual.proximo

        return equipes_regiao

    def simular_dia_emergencias(self, num_ocorrencias: int = 20) -> None:
        """
        Simula um dia completo de emergÃªncias com severidade crescente
        Args:
            num_ocorrencias: NÃºmero de ocorrÃªncias a simular
        """
        print(f"\nğŸ”¥ === SIMULAÃ‡ÃƒO DE {num_ocorrencias} OCORRÃŠNCIAS ===")
        print("ğŸ“ˆ Simulando severidade crescente ao longo do dia...")

        tipos_ocorrencia = [
            "IncÃªndio em VegetaÃ§Ã£o",
            "Queimada Controlada",
            "Foco em Pasto",
            "IncÃªndio Florestal",
            "IncÃªndio Urbano",
            "IncÃªndio Industrial",
            "IncÃªndio em PrÃ©dio",
            "IncÃªndio Residencial",
            "Queimada Rural",
            "IncÃªndio em LixÃ£o"
        ]

        localizacoes = [
            "Zona Norte", "Zona Sul", "Zona Leste", "Zona Oeste",
            "Centro", "Periferia", "Zona Rural", "Distrito Industrial",
            "Reserva Florestal", "Ãrea Urbana"
        ]

        regioes = ["norte", "sul", "leste", "oeste"]

        for i in range(num_ocorrencias):
            # Severidade crescente: primeiras ocorrÃªncias tÃªm prioridade baixa,
            # Ãºltimas tÃªm prioridade alta
            prioridade_base = int((i / num_ocorrencias) * 7) + 1  # 1-8
            prioridade = min(prioridade_base + random.randint(-1, 2), 10)

            descricao = random.choice(tipos_ocorrencia)
            localizacao = random.choice(localizacoes)
            regiao = random.choice(regioes)

            ocorrencia = self.criar_ocorrencia(descricao, prioridade, localizacao, regiao)

            # Pequena pausa entre criaÃ§Ãµes
            time.sleep(0.1)

        print(f"\nâœ… {num_ocorrencias} ocorrÃªncias criadas com severidade crescente!")

        # Processa primeiro as emergÃªncias crÃ­ticas
        self.processar_emergencias_criticas()

        # Depois processa filas regionais
        for regiao in regioes:
            self.processar_fila_regional(regiao)

    def gerar_relatorio_completo(self) -> None:
        """Gera relatÃ³rio completo do sistema"""
        print("\nğŸ“Š === RELATÃ“RIO COMPLETO CURUPIRA SIMULATOR ===")

        # EstatÃ­sticas gerais
        print(f"ğŸ“ˆ ESTATÃSTICAS GERAIS:")
        for chave, valor in self.estatisticas_gerais.items():
            print(f"   ğŸ“‹ {chave.replace('_', ' ').title()}: {valor}")

        # Estado das estruturas
        print(f"\nğŸ”¥ ESTADO DAS ESTRUTURAS:")
        print(f"   ğŸš¨ Heap EmergÃªncias: {self.heap_emergencias.tamanho()} pendentes")
        print(f"   ğŸ“‹ Fila Norte: {self.fila_norte.tamanho()} pendentes")
        print(f"   ğŸ“‹ Fila Sul: {self.fila_sul.tamanho()} pendentes")
        print(f"   ğŸ“‹ Fila Leste: {self.fila_leste.tamanho()} pendentes")
        print(f"   ğŸ“‹ Fila Oeste: {self.fila_oeste.tamanho()} pendentes")

        # EstatÃ­sticas das equipes
        stats_equipes = self.lista_equipes.obter_estatisticas()
        print(f"\nğŸš’ ESTATÃSTICAS DAS EQUIPES:")
        for chave, valor in stats_equipes.items():
            print(f"   ğŸ‘¥ {chave.replace('_', ' ').title()}: {valor}")

        # HistÃ³rico resumido das equipes
        print(f"\nğŸ“š RESUMO DOS HISTÃ“RICOS:")
        for equipe_id, historico in self.historicos.items():
            total_acoes = len(historico.pilha)
            print(f"   ğŸ“ {equipe_id}: {total_acoes} aÃ§Ãµes registradas")

    def visualizar_sistema_completo(self) -> None:
        """Visualiza estado completo de todas as estruturas"""
        print("\nğŸŒ³ === VISUALIZAÃ‡ÃƒO COMPLETA DO SISTEMA ===")

        # Heap de emergÃªncias
        self.heap_emergencias.visualizar_heap()

        # Filas regionais
        for fila in [self.fila_norte, self.fila_sul, self.fila_leste, self.fila_oeste]:
            fila.visualizar_fila()

        # Lista de equipes
        self.lista_equipes.visualizar_lista()

        # Ãrvore de regiÃµes
        self.arvore_regioes.visualizar_arvore()

        # HistÃ³ricos das equipes (Ãºltimas 3 aÃ§Ãµes)
        print(f"\nğŸ“š === ÃšLTIMAS AÃ‡Ã•ES DAS EQUIPES ===")
        for equipe_id, historico in self.historicos.items():
            historico.visualizar_historico(3)

# DemonstraÃ§Ã£o do Sistema Completo
print("ğŸŒ³ === DEMONSTRAÃ‡ÃƒO COMPLETA: CURUPIRA SIMULATOR ===")

# Criando sistema integrado
curupira = CurupiraSimulator()

# Executando simulaÃ§Ã£o completa
curupira.simular_dia_emergencias(15)

# Gerando relatÃ³rios
curupira.gerar_relatorio_completo()
curupira.visualizar_sistema_completo()

print("\nğŸŒ³ === SIMULAÃ‡ÃƒO CONCLUÃDA COM SUCESSO === ğŸŒ³")
print("âœ… Todas as 5 estruturas demonstradas!")
print("ğŸ“Š Sistema Curupira Digital operacional!")
```
