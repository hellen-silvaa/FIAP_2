```python
# ===================================================================
#                    üö® SISTEMA DE FILAS PARA OCORR√äNCIAS
#              Gerenciamento FIFO de Alertas de Queimadas
#
#          * Conjunto 1: 'Fila', Pilha, Lista ligada, √Årvore, Heap
# ===================================================================


print("üö® SISTEMA DE GERENCIAMENTO DE OCORR√äNCIAS - CENTRAL DE BOMBEIROS")
print("üì° Alertas recebidos dos drones Curupira Digital")
print("=" * 80)

fila_ocorrencias = []

def adicionar_ocorrencia():
    """Adiciona nova ocorr√™ncia na fila"""
    print("\nüî• NOVA OCORR√äNCIA DETECTADA PELOS DRONES")

    # usa os munic√≠pios cr√≠ticos da c√©lula anterior
    municipio = random.choice(todos_municipios_criticos)

    # gera dados da ocorr√™ncia baseado nos dados
    info_regiao = regioes_por_municipio[municipio]

    # cria a ocorr√™ncia
    timestamp = f"{random.randint(6, 23):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d}"
    severidade = random.choice(["BAIXA", "M√âDIA", "ALTA", "CR√çTICA"])

    ocorrencia = {
        "id": len(fila_ocorrencias) + 1,
        "municipio": municipio,
        "regiao": info_regiao["regiao"],
        "timestamp": timestamp,
        "severidade": severidade,
        "vegetacao": info_regiao["vegetacao"],
        "status": "AGUARDANDO"
    }

    fila_ocorrencias.append(ocorrencia)
    print(f"üöÅ Ocorr√™ncia #{ocorrencia['id']} adicionada √† fila:")
    print(f"   üìç Local: {municipio}")
    print(f"   ‚ö†Ô∏è Severidade: {severidade}")
    print(f"   üïê Hor√°rio: {timestamp}")

def atender_proxima_ocorrencia():
    """Atende pr√≥xima ocorr√™ncia FIFO"""
    if fila_ocorrencias:
        # fifo- remove sempre o primeiro (pop(0))
        atendida = fila_ocorrencias.pop(0)
        print(f"\nüöí ATENDENDO OCORR√äNCIA #{atendida['id']}")
        print(f"   üìç {atendida['municipio']} - {atendida['regiao']}")
        print(f"   ‚ö†Ô∏è Severidade: {atendida['severidade']}")
        print(f"   üïê Recebida √†s: {atendida['timestamp']}")
        print(f"   üå≥ Vegeta√ß√£o: {atendida['vegetacao']}")
        return atendida
    else:
        print("\n‚úÖ Fila vazia, nenhuma ocorr√™ncia para atender.")
        return None

def mostrar_fila():
    """Mostra fila atual"""
    if fila_ocorrencias:
        print(f"\nüìã FILA ATUAL DE OCORR√äNCIAS ({len(fila_ocorrencias)} aguardando):")
        print("-" * 80)
        for i, ocorrencia in enumerate(fila_ocorrencias, 1):
            print(f"   {i}. #{ocorrencia['id']} | {ocorrencia['municipio']} | "
                  f"{ocorrencia['severidade']} | {ocorrencia['timestamp']}")
    else:
        print("\n‚úÖ Fila vazia.")

# adiciona algumas ocorr√™ncias para teste
print("\nü§ñ SIMULANDO CHEGADA DE ALERTAS DOS DRONES...")
for i in range(5):
    adicionar_ocorrencia()

# menu interativo
print("\n" + "="*80)
print("üéõÔ∏è CENTRAL DE CONTROLE - SISTEMA FIFO")
print("="*80)

while True:
    opcao = input("""
üö® ESCOLHA UMA OP√á√ÉO:
1Ô∏è‚É£ - Adicionar Nova Ocorr√™ncia
2Ô∏è‚É£ - Atender Pr√≥xima Ocorr√™ncia (FIFO)
3Ô∏è‚É£ - Mostrar Fila Atual
4Ô∏è‚É£ - Gerar M√∫ltiplas Ocorr√™ncias
5Ô∏è‚É£ - Sair do Sistema

üëâ Digite sua op√ß√£o: """)

    if opcao == "1":
        adicionar_ocorrencia()

    elif opcao == "2":
        atendida = atender_proxima_ocorrencia()

    elif opcao == "3":
        mostrar_fila()

    elif opcao == "4":
        quantidade = int(input("Quantas ocorr√™ncias gerar? "))
        for _ in range(quantidade):
            adicionar_ocorrencia()
        print(f"\n‚úÖ {quantidade} ocorr√™ncias adicionadas √† fila!")

    elif opcao == "5":
        print("\nüîö Encerrando sistema de gerenciamento...")
        print("üìä Estat√≠sticas finais:")
        print(f"   ‚Ä¢ Ocorr√™ncias pendentes: {len(fila_ocorrencias)}")
        break

    else:
        print("‚ùå Op√ß√£o inv√°lida! Tente novamente.")

```

```python
# ===================================================================
#                    üìö SISTEMA DE PILHAS - HIST√ìRICO DE A√á√ïES
#              Gerenciamento LIFO de A√ß√µes dos Bombeiros
#
#          * Conjunto 1: Fila, 'Pilha', Lista ligada, √Årvore, Heap
# ===================================================================

print("üìö SISTEMA DE HIST√ìRICO DE A√á√ïES - BOMBEIROS")
print("üîÑ √öltima a√ß√£o registrada √© a primeira a ser consultada - LIFO")
print("=" * 80)

pilha_acoes = []

def registrar_acao():
    """Registra nova a√ß√£o na pilha"""
    print("\nüìù REGISTRANDO NOVA A√á√ÉO DOS BOMBEIROS")

    # usa os munic√≠pios cr√≠ticos da c√©lula anterior
    municipio = random.choice(todos_municipios_criticos)

    # tipos de a√ß√µes dos bombeiros
    tipos_acao = [
        "COMBATE DIRETO",
        "RECONHECIMENTO A√âREO",
        "EVACUA√á√ÉO PREVENTIVA",
        "CONTROLE DE PER√çMETRO",
        "RESCALDO",
        "APOIO LOG√çSTICO",
        "COORDENA√á√ÉO LOCAL"
    ]

    # cria a a√ß√£o
    timestamp = f"{random.randint(6, 23):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d}"
    tipo_acao = random.choice(tipos_acao)

    acao = {
        "id": len(pilha_acoes) + 1,
        "tipo": tipo_acao,
        "municipio": municipio,
        "timestamp": timestamp,
        "equipe": f"Equipe-{random.randint(1, 15)}",
        "duracao": f"{random.randint(15, 180)}min",
        "status": "CONCLU√çDA"
    }

    pilha_acoes.append(acao)
    print(f"‚úÖ A√ß√£o #{acao['id']} registrada na pilha:")
    print(f"   üéØ Tipo: {tipo_acao}")
    print(f"   üìç Local: {municipio}")
    print(f"   üë• Equipe: {acao['equipe']}")
    print(f"   ‚è±Ô∏è Dura√ß√£o: {acao['duracao']}")
    print(f"   üïê Hor√°rio: {timestamp}")

def consultar_ultima_acao():
    """Consulta e remove √∫ltima a√ß√£o LIFO """
    if pilha_acoes:
        # lifo remove sempre o √∫ltimo (pop())
        ultima = pilha_acoes.pop()
        print(f"\nüîç CONSULTANDO √öLTIMA A√á√ÉO REGISTRADA #{ultima['id']}")
        print(f"   üéØ Tipo: {ultima['tipo']}")
        print(f"   üìç Local: {ultima['municipio']}")
        print(f"   üë• Equipe: {ultima['equipe']}")
        print(f"   ‚è±Ô∏è Dura√ß√£o: {ultima['duracao']}")
        print(f"   üïê Registrada √†s: {ultima['timestamp']}")
        print(f"   ‚úÖ Status: {ultima['status']}")
        return ultima
    else:
        print("\nüì≠ Pilha vazia, nenhuma a√ß√£o registrada.")
        return None

def mostrar_pilha():
    """Mostra pilha atual"""
    if pilha_acoes:
        print(f"\nüìö HIST√ìRICO ATUAL DE A√á√ïES ({len(pilha_acoes)} registradas):")
        print("   (√öltima registrada aparece primeiro - LIFO)")
        print("-" * 80)
        # mostra do √∫ltimo para o primeiro - LIFO
        for i, acao in enumerate(reversed(pilha_acoes), 1):
            print(f"   {i}. #{acao['id']} | {acao['tipo']} | "
                  f"{acao['municipio']} | {acao['equipe']} | {acao['timestamp']}")
    else:
        print("\nüì≠ Hist√≥rico vazio.")

def mostrar_top_da_pilha():
    """Mostra √∫ltima a√ß√£o SEM remover"""
    if pilha_acoes:
        ultima = pilha_acoes[-1]  # acessa o √∫ltimo elemento sem remover
        print(f"\nüîù √öLTIMA A√á√ÉO REGISTRADA (topo da pilha):")
        print(f"   üéØ #{ultima['id']} - {ultima['tipo']}")
        print(f"   üìç {ultima['municipio']} | {ultima['equipe']}")
        print(f"   üïê {ultima['timestamp']}")
    else:
        print("\nüì≠ Pilha vazia.")

# registra algumas a√ß√µes para teste
print("\nü§ñ SIMULANDO REGISTRO DE A√á√ïES DOS BOMBEIROS...")
for i in range(4):
    registrar_acao()

# menu interativo
print("\n" + "="*80)
print("üìö CENTRAL DE HIST√ìRICO - SISTEMA LIFO")
print("="*80)

while True:
    opcao = input("""
üìö ESCOLHA UMA OP√á√ÉO:
1Ô∏è‚É£ - Registrar Nova A√ß√£o
2Ô∏è‚É£ - Consultar √öltima A√ß√£o (LIFO)
3Ô∏è‚É£ - Mostrar Hist√≥rico Completo
4Ô∏è‚É£ - Ver Topo da Pilha (sem remover)
5Ô∏è‚É£ - Registrar M√∫ltiplas A√ß√µes
6Ô∏è‚É£ - Sair do Sistema

üëâ Digite sua op√ß√£o: """)

    if opcao == "1":
        registrar_acao()

    elif opcao == "2":
        consultada = consultar_ultima_acao()

    elif opcao == "3":
        mostrar_pilha()

    elif opcao == "4":
        mostrar_top_da_pilha()

    elif opcao == "5":
        quantidade = int(input("Quantas a√ß√µes registrar? "))
        for _ in range(quantidade):
            registrar_acao()
        print(f"\n‚úÖ {quantidade} a√ß√µes registradas na pilha!")

    elif opcao == "6":
        print("\nüîö Encerrando sistema de hist√≥rico...")
        print("üìä Estat√≠sticas finais:")
        print(f"   ‚Ä¢ A√ß√µes no hist√≥rico: {len(pilha_acoes)}")
        break

    else:
        print("‚ùå Op√ß√£o inv√°lida! Tente novamente.")

print("üîÑ √öltima a√ß√£o registrada sempre fica no topo da pilha!")
```

```python
# ===================================================================
#         üöÅ SISTEMA DE LISTA LIGADA - CADEIA DE DRONES
#Estrutura Encadeada de Drones em Opera√ß√£o do sistema Curupira Digital
#
#          * Conjunto 1: Fila, Pilha, 'Lista ligada', √Årvore, Heap
# ===================================================================

print(" SISTEMA DE LISTA LIGADA - CADEIA DE DRONES CURUPIRA")
print(" Cada drone conectado ao pr√≥ximo formando uma cadeia de vigil√¢ncia")
print("=" * 80)

# classe n√≥- estrutura b√°sica da lista ligada
class NodeDrone:
    """N√≥ da lista ligada representando um drone"""
    def __init__(self, dados_drone):
        self.dados = dados_drone  # informa√ß√µes do drone
        self.proximo = None       # refer√™ncia para o pr√≥ximo drone

# classe lista ligada - cadeia de drones
class CadeiaDrones:
    """Lista ligada para gerenciar cadeia de drones"""
    def __init__(self):
        self.cabeca = None  # 1¬∞ drone da cadeia
        self.tamanho = 0

    def adicionar_drone(self, id_drone, municipio, status="ATIVO"):
        """adiciona novo drone no in√≠cio da cadeia"""
        dados_drone = {
            "id": id_drone,
            "municipio": municipio,
            "status": status,
            "timestamp": f"{random.randint(6, 23):02d}:{random.randint(0, 59):02d}:{random.randint(0, 59):02d}",
            "bateria": random.randint(45, 100),
            "altitude": random.randint(50, 200)
        }

        # cria novo n√≥
        novo_drone = NodeDrone(dados_drone)

        # conecta na cadeia
        novo_drone.proximo = self.cabeca
        self.cabeca = novo_drone
        self.tamanho += 1

        print(f"üöÅ Drone {id_drone} adicionado √† cadeia:")
        print(f"   üìç Patrulhando: {municipio}")
        print(f"   üîã Bateria: {dados_drone['bateria']}%")
        print(f"   ‚úàÔ∏è Altitude: {dados_drone['altitude']}m")
        print(f"   üïê Ativado √†s: {dados_drone['timestamp']}")

    def remover_drone(self, id_drone):
        """remove drone espec√≠fico da cadeia"""
        if not self.cabeca:
            print("‚ùå Cadeia vazia - nenhum drone para remover")
            return False

        # se √© o primeiro da cadeia
        if self.cabeca.dados["id"] == id_drone:
            removido = self.cabeca.dados
            self.cabeca = self.cabeca.proximo
            self.tamanho -= 1
            print(f"üöÅ Drone {id_drone} removido da cadeia")
            print(f"   üìç Estava em: {removido['municipio']}")
            return True

        # procura na cadeia
        atual = self.cabeca
        while atual.proximo:
            if atual.proximo.dados["id"] == id_drone:
                removido = atual.proximo.dados
                atual.proximo = atual.proximo.proximo
                self.tamanho -= 1
                print(f"üöÅ Drone {id_drone} removido da cadeia")
                print(f"   üìç Estava em: {removido['municipio']}")
                return True
            atual = atual.proximo

        print(f"‚ùå Drone {id_drone} n√£o encontrado na cadeia")
        return False

    def buscar_drone(self, id_drone):
        """busca drone espec√≠fico na cadeia"""
        atual = self.cabeca
        posicao = 1

        while atual:
            if atual.dados["id"] == id_drone:
                print(f"üîç Drone {id_drone} encontrado na posi√ß√£o {posicao}:")
                print(f"   üìç Local: {atual.dados['municipio']}")
                print(f"   üîã Bateria: {atual.dados['bateria']}%")
                print(f"   üì° Status: {atual.dados['status']}")
                return atual.dados
            atual = atual.proximo
            posicao += 1

        print(f"‚ùå Drone {id_drone} n√£o encontrado na cadeia")
        return None

    def mostrar_cadeia(self):
        """mostra toda a cadeia de drones conectados"""
        if not self.cabeca:
            print("üì≠ Cadeia vazia - nenhum drone em opera√ß√£o")
            return

        print(f"\nüîó CADEIA ATUAL DE DRONES ({self.tamanho} conectados):")
        print("-" * 80)

        atual = self.cabeca
        posicao = 1

        while atual:
            drone = atual.dados
            conexao = " ‚Üí " if atual.proximo else " ‚Üí [FIM]"

            print(f"   {posicao}. Drone-{drone['id']} | {drone['municipio']} | "
                  f"üîã{drone['bateria']}% | {drone['status']}{conexao}")

            atual = atual.proximo
            posicao += 1

    def atualizar_status_drone(self, id_drone, novo_status):
        """atualiza o status de um drone na cadeia"""
        atual = self.cabeca

        while atual:
            if atual.dados["id"] == id_drone:
                status_anterior = atual.dados["status"]
                atual.dados["status"] = novo_status
                print(f"üîÑ Status do Drone {id_drone} atualizado:")
                print(f"   üìç Local: {atual.dados['municipio']}")
                print(f"   üîÑ {status_anterior} ‚Üí {novo_status}")
                return True
            atual = atual.proximo

        print(f"‚ùå Drone {id_drone} n√£o encontrado para atualiza√ß√£o")
        return False

    def drones_por_status(self, status):
        """lista todos os drones com determinado status"""
        drones_encontrados = []
        atual = self.cabeca

        while atual:
            if atual.dados["status"] == status:
                drones_encontrados.append(atual.dados)
            atual = atual.proximo

        if drones_encontrados:
            print(f"\nüìä DRONES COM STATUS '{status}' ({len(drones_encontrados)}):")
            for drone in drones_encontrados:
                print(f"   üöÅ Drone-{drone['id']} em {drone['municipio']} | üîã{drone['bateria']}%")
        else:
            print(f"‚ùå Nenhum drone com status '{status}' encontrado")

        return drones_encontrados

# criando cadeia de drones
cadeia_vigilancia = CadeiaDrones()

drones_iniciais = [
    ("D001", "Colniza"),
    ("D002", "Aripuan√£"),
    ("D003", "Nova Bandeirantes"),
    ("D004", "Formoso do Araguaia"),
    ("D005", "Lagoa da Confus√£o"),
    ("D006", "Barreiras"),
    ("D007", "S√£o Desid√©rio")
]

print("\nü§ñ ADICIONANDO DRONES √Ä CADEIA...")
for id_drone, municipio in drones_iniciais:
    cadeia_vigilancia.adicionar_drone(id_drone, municipio)

# demonstrando opera√ß√£o
print("\n" + "="*80)
print("üîó DEMONSTRA√á√ÉO DA LISTA LIGADA")
print("="*80)

# mostra a cadeia inicial
cadeia_vigilancia.mostrar_cadeia()

# busca por um drone espec√≠fico
print("\nüîç BUSCANDO DRONE D003...")
cadeia_vigilancia.buscar_drone("D003")

print("\nüîÑ ATUALIZANDO STATUS DE DRONES...")
cadeia_vigilancia.atualizar_status_drone("D002", "MANUTEN√á√ÉO")
cadeia_vigilancia.atualizar_status_drone("D005", "RETORNANDO")

cadeia_vigilancia.drones_por_status("ATIVO")

# remove drone
print("\nüöÅ REMOVENDO DRONE D004 DA CADEIA...")
cadeia_vigilancia.remover_drone("D004")

# mostra cadeia final
print("\nüîó CADEIA AP√ìS MODIFICA√á√ïES:")
cadeia_vigilancia.mostrar_cadeia()

print(f"\nüìä ESTAT√çSTICAS DA CADEIA:")
print(f"  Total de drones conectados: {cadeia_vigilancia.tamanho}")
print(f"  Estrutura: Lista Ligada (cada n√≥ aponta para o pr√≥ximo)")
print(f"  Vantagem: Inser√ß√£o/remo√ß√£o eficiente no in√≠cio da cadeia")
```

```python
# ===================================================================
#           üå≥ SISTEMA DE √ÅRVORE BIN√ÅRIA - HIERARQUIA DE REGI√ïES
#              Organiza√ß√£o Hier√°rquica de Comando e Controle
#
#          * Conjunto 1: Fila, Pilha, Lista ligada, '√Årvore', Heap
# ===================================================================

print("üóÇÔ∏è Organiza√ß√£o hier√°rquica das regi√µes por n√≠vel de criticidade")
print("=" * 80)

# classe n√≥ da √°rvore - estrutura b√°sica
class NodeRegiao:
    """n√≥ da √°rvore bin√°ria representando uma regi√£o"""
    def __init__(self, dados_regiao):
        self.dados = dados_regiao
        self.esquerda = None  # filhos √† esquerda (menor criticidade)
        self.direita = None   # filhos √† direita (maior criticidade)

# classe √°rvore binaria - hierarquia de regi√µes
class ArvoreRegioes:
    """√°rvore bin√°ria para organizar regi√µes por criticidade"""
    def __init__(self):
        self.raiz = None

    def inserir_regiao(self, nome_regiao, nivel_criticidade, estados, focos_totais):
        """Insere nova regi√£o na √°rvore baseado na criticidade"""
        dados_regiao = {
            "nome": nome_regiao,
            "criticidade": nivel_criticidade,
            "estados": estados,
            "focos": focos_totais,
            "timestamp": f"{random.randint(6, 23):02d}:{random.randint(0, 59):02d}",
            "status_operacional": "MONITORANDO"
        }

        if self.raiz is None:
            self.raiz = NodeRegiao(dados_regiao)
            print(f"üå≥ Regi√£o RAIZ criada: {nome_regiao}")
        else:
            self._inserir_recursivo(self.raiz, dados_regiao)
            print(f"üåø Regi√£o {nome_regiao} inserida na √°rvore")

        print(f"   üìä Criticidade: {nivel_criticidade}")
        print(f"   üî• Focos: {focos_totais}")
        print(f"   üìç Estados: {', '.join(estados)}")

    def _inserir_recursivo(self, no_atual, novos_dados):
        """Fun√ß√£o recursiva para inserir na posi√ß√£o correta"""
        # se a criticidade for menor, vai para esquerda
        if novos_dados["focos"] < no_atual.dados["focos"]:
            if no_atual.esquerda is None:
                no_atual.esquerda = NodeRegiao(novos_dados)
            else:
                self._inserir_recursivo(no_atual.esquerda, novos_dados)
        # se for maior ou igual, vai para direita
        else:
            if no_atual.direita is None:
                no_atual.direita = NodeRegiao(novos_dados)
            else:
                self._inserir_recursivo(no_atual.direita, novos_dados)

    def buscar_regiao(self, nome_regiao):
        """Busca regi√£o espec√≠fica na √°rvore"""
        print(f"üîç Buscando regi√£o: {nome_regiao}")
        resultado = self._buscar_recursivo(self.raiz, nome_regiao)

        if resultado:
            print(f"‚úÖ Regi√£o encontrada:")
            print(f"   üìä Criticidade: {resultado['criticidade']}")
            print(f"   üî• Focos: {resultado['focos']}")
            print(f"   üìç Estados: {', '.join(resultado['estados'])}")
            return resultado
        else:
            print(f"‚ùå Regi√£o {nome_regiao} n√£o encontrada")
            return None

    def _buscar_recursivo(self, no_atual, nome_procurado):
        """Fun√ß√£o recursiva para buscar regi√£o"""
        if no_atual is None:
            return None

        if no_atual.dados["nome"] == nome_procurado:
            return no_atual.dados

        # busca nos dois lados
        resultado_esq = self._buscar_recursivo(no_atual.esquerda, nome_procurado)
        if resultado_esq:
            return resultado_esq

        return self._buscar_recursivo(no_atual.direita, nome_procurado)

    def percorrer_em_ordem(self):
        """Percorre √°rvore em ordem (esquerda ‚Üí raiz ‚Üí direita)"""
        print("\nüå≥ PERCORRENDO √ÅRVORE EM ORDEM (Menor ‚Üí Maior criticidade):")
        print("-" * 80)
        self._em_ordem_recursivo(self.raiz)

    def _em_ordem_recursivo(self, no_atual):
        """Fun√ß√£o recursiva para percorrer em ordem"""
        if no_atual is not None:
            # primeiro visita sub√°rvore esquerda
            self._em_ordem_recursivo(no_atual.esquerda)

            # depois processa n√≥ atual
            regiao = no_atual.dados
            print(f"   üåø {regiao['nome']} | Focos: {regiao['focos']} | "
                  f"Criticidade: {regiao['criticidade']}")

            # por √∫ltimo visita sub√°rvore direita
            self._em_ordem_recursivo(no_atual.direita)

    def percorrer_pre_ordem(self):
        """percorre √°rvore em pr√©-ordem (raiz ‚Üí esquerda ‚Üí direita)"""
        print("\nüå≥ PERCORRENDO √ÅRVORE EM PR√â-ORDEM (Hierarquia de comando):")
        print("-" * 80)
        self._pre_ordem_recursivo(self.raiz, 0)

    def _pre_ordem_recursivo(self, no_atual, nivel):
        """fun√ß√£o recursiva para percorrer em pr√©-ordem"""
        if no_atual is not None:
            # primeiro processa n√≥ atual (com indenta√ß√£o por n√≠vel)
            indent = "   " * nivel
            regiao = no_atual.dados
            print(f"{indent}üåø {regiao['nome']} | Focos: {regiao['focos']} | "
                  f"Status: {regiao['status_operacional']}")

            # depois visita filhos
            self._pre_ordem_recursivo(no_atual.esquerda, nivel + 1)
            self._pre_ordem_recursivo(no_atual.direita, nivel + 1)

    def contar_regioes(self):
        """conta total de regi√µes na √°rvore"""
        return self._contar_recursivo(self.raiz)

    def _contar_recursivo(self, no_atual):
        """fun√ß√£o recursiva para contar n√≥s"""
        if no_atual is None:
            return 0
        return 1 + self._contar_recursivo(no_atual.esquerda) + self._contar_recursivo(no_atual.direita)

    def atualizar_status_regiao(self, nome_regiao, novo_status):
        """atualiza status operacional de uma regi√£o"""
        no_encontrado = self._encontrar_no(self.raiz, nome_regiao)

        if no_encontrado:
            status_anterior = no_encontrado.dados["status_operacional"]
            no_encontrado.dados["status_operacional"] = novo_status
            print(f"üîÑ Status da regi√£o {nome_regiao} atualizado:")
            print(f"   üìä {status_anterior} ‚Üí {novo_status}")
            return True
        else:
            print(f"‚ùå Regi√£o {nome_regiao} n√£o encontrada para atualiza√ß√£o")
            return False

    def _encontrar_no(self, no_atual, nome_procurado):
        """fun√ß√£o auxiliar para encontrar n√≥ espec√≠fico"""
        if no_atual is None:
            return None

        if no_atual.dados["nome"] == nome_procurado:
            return no_atual

        # busca nos dois lados
        resultado_esq = self._encontrar_no(no_atual.esquerda, nome_procurado)
        if resultado_esq:
            return resultado_esq

        return self._encontrar_no(no_atual.direita, nome_procurado)

# cria √°rvore de regi√µes
print("üöÄ INICIALIZANDO √ÅRVORE HIER√ÅRQUICA DE REGI√ïES...")
arvore_comando = ArvoreRegioes()

# dados das regi√µes (usando dados da primeira c√©lula)
regioes_hierarquia = [
    ("Centro_Oeste_Norte", "EXTREMO", ["MT", "TO", "MA"], 4792),
    ("Nordeste_Cerrado", "ALTO", ["BA", "PI"], 2049),
    ("Amaz√¥nia_Norte", "CR√çTICO", ["RR", "PA"], 1712),
    ("Regi√£o_Sul_Pantanal", "MODERADO", ["RS", "MS", "MG"], 1819)
]

# inserindo regi√µes na √°rvore
print("\nüå≥ CONSTRUINDO √ÅRVORE HIER√ÅRQUICA...")
for nome, criticidade, estados, focos in regioes_hierarquia:
    arvore_comando.inserir_regiao(nome, criticidade, estados, focos)

# demonstra√ß√£o das opera√ß~eos da √°rvore
print("\n" + "="*80)
print("üå≥ DEMONSTRA√á√ÉO DA √ÅRVORE BIN√ÅRIA")
print("="*80)

# percorre √°rvore em diferentes ordens
arvore_comando.percorrer_em_ordem()
arvore_comando.percorrer_pre_ordem()

# busca regi√£o espec√≠fica
print("\nüîç BUSCANDO REGI√ÉO ESPEC√çFICA...")
arvore_comando.buscar_regiao("Centro_Oeste_Norte")

# atualiza status
print("\nüîÑ ATUALIZANDO STATUS OPERACIONAL...")
arvore_comando.atualizar_status_regiao("Amaz√¥nia_Norte", "EM EMERG√äNCIA")
arvore_comando.atualizar_status_regiao("Regi√£o_Sul_Pantanal", "STANDBY")

# estat√≠sticas da √°rvore
total_regioes = arvore_comando.contar_regioes()

print(f"\nüìä ESTAT√çSTICAS DA √ÅRVORE:")
print(f"   üå≥ Total de regi√µes na hierarquia: {total_regioes}")
print(f"   üóÇÔ∏è Estrutura: √Årvore Bin√°ria de Busca")
print(f"   üìä Organiza√ß√£o: Por n√∫mero de focos (menor ‚Üê raiz ‚Üí maior)")
print(f"   ‚ö° Vantagem: Busca e inser√ß√£o eficientes O(log n)")

# mostra estrutura final
print("\nüå≥ ESTRUTURA HIER√ÅRQUICA FINAL:")
arvore_comando.percorrer_pre_ordem()

print("\nüå≥ √Årvore Bin√°ria implementada para hierarquia de comando!")
```

```python
# ===================================================================
#                    üö® SISTEMA DE HEAP - PRIORIZA√á√ÉO DE EMERG√äNCIAS
#              Fila de Prioridade para Ocorr√™ncias Cr√≠ticas
#
#          * Conjunto 1: Fila, Pilha, Lista ligada, √Årvore, 'Heap'
# ===================================================================

import heapq

print("üö® SISTEMA DE HEAP - PRIORIZA√á√ÉO DE EMERG√äNCIAS CURUPIRA")
print("‚ö° Fila de prioridade: ocorr√™ncias mais cr√≠ticas atendidas primeiro")
print("=" * 80)

class HeapPrioridades:
    """üìä Heap para gerenciar prioridades de ocorr√™ncias"""

    def __init__(self):
        self.heap = []  # MinHeap (menor prioridade sai primeiro)
        self.contador = 0  # Para evitar empates

    def calcular_prioridade(self, severidade, municipio, tipo_causa):
        """üìä Calcula prioridade baseada em m√∫ltiplos fatores"""
        # prioridade base pela severidade (1-10)
        prioridade = severidade

        # b√¥nus por tipo de causa
        bonus_causa = {
            "criminosa": 3.0,
            "desmatamento": 2.5,
            "acidental": 1.5,
            "natural": 1.0,
            "raio": 1.2
        }
        prioridade += bonus_causa.get(tipo_causa, 1.0)

        # b√¥nus por munic√≠pio cr√≠tico
        municipios_criticos = ["Colniza", "Barreiras", "Formoso do Araguaia", "Rorain√≥polis"]
        if municipio in municipios_criticos:
            prioridade += 2.0

        return round(prioridade, 2)

    def inserir_ocorrencia(self, municipio, severidade, tipo_causa, observacoes=""):
        """üìà Insere nova ocorr√™ncia no heap"""

        # calcula prioridade
        prioridade = self.calcular_prioridade(severidade, municipio, tipo_causa)

        # cria ocorr√™ncia
        self.contador += 1
        ocorrencia = {
            'id': self.contador,
            'municipio': municipio,
            'severidade': severidade,
            'tipo_causa': tipo_causa,
            'observacoes': observacoes,
            'prioridade': prioridade,
            'timestamp': f"{random.randint(6, 23):02d}:{random.randint(0, 59):02d}",
            'status': 'AGUARDANDO'
        }

        # usamos negativo para simular MaxHeap , maior prioridade primeiro
        prioridade_negativa = -prioridade

        # tupla: prioridade, contador, ocorr√™ncia
        item_heap = (prioridade_negativa, self.contador, ocorrencia)
        heapq.heappush(self.heap, item_heap)

        print(f"üìà HEAP: Ocorr√™ncia #{ocorrencia['id']} inserida")
        print(f"   üìç Munic√≠pio: {municipio}")
        print(f"   üî• Severidade: {severidade}/10")
        print(f"   ‚ö° Prioridade: {prioridade}")
        print(f"   üìä Tamanho do heap: {len(self.heap)}")

        return self.contador

    def remover_maior_prioridade(self):
        """üìâ Remove ocorr√™ncia com MAIOR prioridade do heap"""

        if not self.heap:
            print("‚ùå HEAP: Vazio! Nenhuma ocorr√™ncia para atender")
            return None

        # remove menor valor, que √© a maior prioridade por estar negativa
        prioridade_negativa, contador, ocorrencia = heapq.heappop(self.heap)
        prioridade_real = -prioridade_negativa

        print(f"üìâ HEAP: Removendo ocorr√™ncia de MAIOR prioridade")
        print(f"   üö® Ocorr√™ncia: #{ocorrencia['id']}")
        print(f"   ‚ö° Prioridade: {prioridade_real}")
        print(f"   üìç Munic√≠pio: {ocorrencia['municipio']}")
        print(f"   üìä Tamanho do heap: {len(self.heap)}")

        return ocorrencia

    def consultar_maior_prioridade(self):
        """üëÅÔ∏è Consulta maior prioridade sem remover"""

        if not self.heap:
            print("üëÅÔ∏è HEAP: Vazio!")
            return None

        prioridade_negativa, contador, ocorrencia = self.heap[0]
        prioridade_real = -prioridade_negativa

        print(f"üëÅÔ∏è HEAP: Maior prioridade atual")
        print(f"   üö® Ocorr√™ncia: #{ocorrencia['id']}")
        print(f"   ‚ö° Prioridade: {prioridade_real}")
        print(f"   üìç Munic√≠pio: {ocorrencia['municipio']}")
        print(f"   üî• Severidade: {ocorrencia['severidade']}/10")

        return ocorrencia

    def listar_heap_ordenado(self):
        """üìä Lista heap em ordem de prioridade (sem remover)"""

        if not self.heap:
            print("üìä HEAP: Vazio!")
            return

        # cria c√≥pia do heap para ordenar sem destruir original
        heap_copia = self.heap.copy()
        lista_ordenada = []

        while heap_copia:
            prioridade_neg, contador, occ = heapq.heappop(heap_copia)
            prioridade_real = -prioridade_neg
            lista_ordenada.append((prioridade_real, occ))

        print(f"üìä HEAP ORDENADO POR PRIORIDADE - {len(lista_ordenada)} ocorr√™ncia(s)")
        print("üéØ Ordem de atendimento (maior prioridade primeiro):")
        print("=" * 90)
        print(f"{'ORDEM':<6} {'ID':<4} {'MUNIC√çPIO':<20} {'PRIORIDADE':<11} {'SEVERIDADE':<10} {'CAUSA':<15}")
        print("-" * 90)

        for i, (prioridade, occ) in enumerate(lista_ordenada, 1):
            simbolo = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}¬∞"
            print(f"{simbolo:<6} #{occ['id']:<3} {occ['municipio']:<20} "
                  f"{prioridade:<11.2f} {occ['severidade']:<10} {occ['tipo_causa']:<15}")

# criando sistema de prioridades
print("üöÄ INICIALIZANDO SISTEMA DE HEAP...")
sistema_heap = HeapPrioridades()

print("\n" + "="*80)
print("üß™ DEMONSTRA√á√ÉO DO HEAP DE PRIORIDADES")
print("="*80)

print("\nüìà 1. INSERINDO OCORR√äNCIAS COM DIFERENTES PRIORIDADES:")

# dados de teste usando munic√≠pios da primeira c√©lula
ocorrencias_teste = [
    ("Colniza", 8, "desmatamento", "√Årea cr√≠tica de MT"),
    ("Barreiras", 4, "natural", "Foco pequeno control√°vel"),
    ("Rorain√≥polis", 9, "criminosa", "Inc√™ndio intencional"),
    ("Formoso do Araguaia", 6, "acidental", "Queima agr√≠cola"),
    ("Corumb√°", 3, "natural", "Monitoramento preventivo"),
    ("A√ßail√¢ndia", 10, "criminosa", "Emerg√™ncia m√°xima")
]

for municipio, severidade, causa, obs in ocorrencias_teste:
    print(f"\nüì• Inserindo: {municipio} (severidade {severidade}, {causa})")
    sistema_heap.inserir_ocorrencia(municipio, severidade, causa, obs)

print("\nüéØ 2. VISUALIZANDO HEAP ORDENADO:")
sistema_heap.listar_heap_ordenado()

print("\nüëÅÔ∏è 3. CONSULTANDO MAIOR PRIORIDADE (sem remover):")
sistema_heap.consultar_maior_prioridade()

print("\nüö® 4. ATENDENDO POR ORDEM DE PRIORIDADE:")

# atende 3 ocorr√™ncias para mostrar funcionamento
for i in range(3):
    print(f"\n--- ATENDIMENTO #{i+1} ---")
    ocorrencia_atendida = sistema_heap.remover_maior_prioridade()

    if i < 2:  # mostra estado ap√≥s cada atendimento
        print(f"\nüìä Estado do heap ap√≥s atendimento {i+1}:")
        sistema_heap.listar_heap_ordenado()

print("\nüìä 5. ESTADO FINAL DO HEAP:")
sistema_heap.listar_heap_ordenado()

print(f"\n‚úÖ HEAP IMPLEMENTADO COM SUCESSO!")
print(f"üéØ Caracter√≠sticas demonstradas:")
print(f"   ‚Ä¢ Inser√ß√£o por prioridade em O(log n)")
print(f"   ‚Ä¢ Remo√ß√£o da maior prioridade em O(log n)")
print(f"   ‚Ä¢ Consulta do m√°ximo em O(1)")
print(f"   ‚Ä¢ Heap mant√©m propriedade sempre balanceada")
```

```python
# ===================================================================
#           üß† SISTEMA DE MEMOIZA√á√ÉO - AN√ÅLISE INTELIGENTE
#              Fun√ß√µes Recursivas com Cache para Otimiza√ß√£o
#
#             Conjunto 2: 'Fun√ß√µes recursivas e memoriza√ß√£o'
# ===================================================================

print("üß† SISTEMA DE MEMOIZA√á√ÉO - AN√ÅLISE INTELIGENTE CURUPIRA")
print("‚ö° Fun√ß√µes recursivas otimizadas com cache para c√°lculos complexos")
print("=" * 80)

class AnaliseInteligente:
    """üß† Sistema de an√°lise com memoiza√ß√£o para otimizar c√°lculos"""

    def __init__(self):
        # caches para memoiza√ß√£o (armazenam resultados j√° calculados)
        self.cache_risco_propagacao = {}
        self.cache_rota_otima = {}
        self.cache_fibonacci_recursos = {}
        self.cache_potencia_fogo = {}

        # contadores para demonstrar efici√™ncia
        self.contador_calculos_risco = 0
        self.contador_calculos_rota = 0
        self.contador_calculos_fibonacci = 0
        self.contador_calculos_potencia = 0

        print("üß† Sistema de memoiza√ß√£o inicializado")
        print("üíæ Caches criados para otimizar c√°lculos recursivos")

    # ================================================================
    #                    FUN√á√ÉO RECURSIVA 1: RISCO DE PROPAGA√á√ÉO
    # ================================================================

    def calcular_risco_propagacao(self, municipio, dias_projecao, vento=5):
        """
        calcula risco de propaga√ß√£o usando recurs√£o com memoiza√ß√£o
        Args:
            municipio: nome do munic√≠pio
            dias_projecao: quantos dias para frente calcular
            vento: velocidade do vento (km/h)
        """
        # cria uma chave √∫nica para o cache
        chave_cache = (municipio, dias_projecao, vento)

        # memoriza√ß√£o: verifica se j√° calculou antes
        if chave_cache in self.cache_risco_propagacao:
            print(f"üíæ CACHE HIT: Risco de {municipio} para {dias_projecao} dias (j√° calculado)")
            return self.cache_risco_propagacao[chave_cache]

        # se n√£o est√° no cache, calcula recursivamente
        self.contador_calculos_risco += 1
        print(f"üßÆ CALCULANDO: Risco de propaga√ß√£o em {municipio} para {dias_projecao} dias")

        # caso base da recurs√£o
        if dias_projecao <= 0:
            return 0.0

        if dias_projecao == 1:
            risco_base = self._obter_risco_base_municipio(municipio)
            resultado = risco_base * (1 + vento/100)

            # memoriza√ß√£o: Armazena no cache
            self.cache_risco_propagacao[chave_cache] = resultado
            return resultado

        # caso recursivo: risco atual + risco dos pr√≥ximos dias (recurs√£o)
        risco_hoje = self._obter_risco_base_municipio(municipio) * (1 + vento/100)
        risco_futuro = self.calcular_risco_propagacao(municipio, dias_projecao - 1, vento * 0.9)

        resultado = risco_hoje + (risco_futuro * 0.8)  # decaimento temporal

        # memoriza√ß√£o: Armazena resultado no cache
        self.cache_risco_propagacao[chave_cache] = resultado

        return resultado

    def _obter_risco_base_municipio(self, municipio):
        """üìä Obt√©m risco base do munic√≠pio (simulado)"""
        riscos_base = {
            "Colniza": 8.5,
            "Barreiras": 6.2,
            "Formoso do Araguaia": 7.8,
            "Rorain√≥polis": 7.1,
            "Corumb√°": 4.3,
            "A√ßail√¢ndia": 8.9,
            "Aripuan√£": 7.6,
            "S√£o Desid√©rio": 5.9
        }
        return riscos_base.get(municipio, 5.0)

    # ================================================================
    #                    FUN√á√ÉO RECURSIVA 2: FIBONACCI RECURSOS
    # ================================================================

    def calcular_recursos_fibonacci(self, nivel_emergencia):
        """
        calcula recursos necess√°rios usando sequ√™ncia de Fibonacci com memoiza√ß√£o
        Simula que recursos crescem em progress√£o de Fibonacci conforme n√≠vel de emerg√™ncia
        """

        # memoriza√ß√£o: Verifica cache
        if nivel_emergencia in self.cache_fibonacci_recursos:
            print(f"üíæ CACHE HIT: Fibonacci recursos n√≠vel {nivel_emergencia}")
            return self.cache_fibonacci_recursos[nivel_emergencia]

        self.contador_calculos_fibonacci += 1
        print(f"üßÆ CALCULANDO: Fibonacci recursos para n√≠vel {nivel_emergencia}")

        # casos base da recurs√£o
        if nivel_emergencia <= 0:
            return 0
        if nivel_emergencia == 1:
            self.cache_fibonacci_recursos[nivel_emergencia] = 1
            return 1
        if nivel_emergencia == 2:
            self.cache_fibonacci_recursos[nivel_emergencia] = 1
            return 1

        # caso recursivo: F(n) = F(n-1) + F(n-2)
        resultado = (self.calcular_recursos_fibonacci(nivel_emergencia - 1) +
                    self.calcular_recursos_fibonacci(nivel_emergencia - 2))

        # memoriza√ß√£o: Armazena no cache
        self.cache_fibonacci_recursos[nivel_emergencia] = resultado

        return resultado

    # ================================================================
    #                    FUN√á√ÉO RECURSIVA 3: POT√äNCIA DE COMBATE
    # ================================================================

    def calcular_potencia_combate(self, base, expoente):
        """
        calcula pot√™ncia de combate usando recurs√£o com memoiza√ß√£o
        simula c√°lculo de for√ßa necess√°ria: base^expoente
        """

        chave_cache = (base, expoente)

        # memoriza√ß√£o: Verifica cache
        if chave_cache in self.cache_potencia_fogo:
            print(f"üíæ CACHE HIT: Pot√™ncia {base}^{expoente}")
            return self.cache_potencia_fogo[chave_cache]

        self.contador_calculos_potencia += 1
        print(f"üßÆ CALCULANDO: Pot√™ncia {base}^{expoente}")

        if expoente == 0:
            resultado = 1
        elif expoente == 1:
            resultado = base
        else:
            # caso recursivo: base^n = base * base^(n-1)
            resultado = base * self.calcular_potencia_combate(base, expoente - 1)

        # memoriza√ß√£o: Armazena no cache
        self.cache_potencia_fogo[chave_cache] = resultado

        return resultado

    # ================================================================
    #                    FUN√á√ÉO RECURSIVA 4: AN√ÅLISE HIER√ÅRQUICA
    # ================================================================

    def analisar_hierarquia_regioes(self, regiao, profundidade_maxima, profundidade_atual=0):
        """
        an√°lise recursiva de hierarquia de regi√µes COM memoiza√ß√£o
        """

        chave_cache = (regiao, profundidade_maxima, profundidade_atual)

        # memoriza√ß√£o
        if chave_cache in self.cache_rota_otima:
            print(f"üíæ CACHE HIT: An√°lise hier√°rquica {regiao} (prof. {profundidade_atual})")
            return self.cache_rota_otima[chave_cache]

        self.contador_calculos_rota += 1

        # caso base
        if profundidade_atual >= profundidade_maxima:
            resultado = {"regiao": regiao, "criticidade": self._obter_criticidade_regiao(regiao)}
            self.cache_rota_otima[chave_cache] = resultado
            return resultado

        # caso recursivo: analisa sub-regi√µes
        sub_regioes = self._obter_sub_regioes(regiao)
        analise_completa = {
            "regiao": regiao,
            "profundidade": profundidade_atual,
            "criticidade": self._obter_criticidade_regiao(regiao),
            "sub_regioes": []
        }

        for sub_regiao in sub_regioes:
            analise_sub = self.analisar_hierarquia_regioes(
                sub_regiao, profundidade_maxima, profundidade_atual + 1
            )
            analise_completa["sub_regioes"].append(analise_sub)

        # memoriza√ß√£o
        self.cache_rota_otima[chave_cache] = analise_completa

        return analise_completa

    def _obter_criticidade_regiao(self, regiao):
        """üìä Simula criticidade de regi√µes"""
        criticidades = {
            "Centro_Oeste": 9.2,
            "Nordeste": 7.5,
            "Norte": 8.1,
            "Sul": 5.3,
            "MT": 8.8,
            "TO": 8.5,
            "BA": 7.2,
            "RR": 7.9
        }
        return criticidades.get(regiao, 5.0)

    def _obter_sub_regioes(self, regiao):
        """üó∫Ô∏è Simula hierarquia de regi√µes"""
        hierarquia = {
            "Centro_Oeste": ["MT", "TO"],
            "Nordeste": ["BA", "PI"],
            "Norte": ["RR", "PA"],
            "Sul": ["RS", "MS"]
        }
        return hierarquia.get(regiao, [])

    # ================================================================
    #                    M√âTODOS DE AN√ÅLISE E ESTAT√çSTICAS
    # ================================================================

    def limpar_caches(self):
        """limpa todos os caches"""
        self.cache_risco_propagacao.clear()
        self.cache_rota_otima.clear()
        self.cache_fibonacci_recursos.clear()
        self.cache_potencia_fogo.clear()

        print("üßπ Todos os caches foram limpos")

    def mostrar_estatisticas_cache(self):
        """üìä Mostra estat√≠sticas dos caches e efici√™ncia"""
        print(f"\nüìä ESTAT√çSTICAS DE memoriza√ß√£o:")
        print("=" * 60)
        print(f"üßÆ Total de c√°lculos realizados:")
        print(f"   ‚Ä¢ Risco de propaga√ß√£o: {self.contador_calculos_risco}")
        print(f"   ‚Ä¢ Fibonacci recursos: {self.contador_calculos_fibonacci}")
        print(f"   ‚Ä¢ Pot√™ncia de combate: {self.contador_calculos_potencia}")
        print(f"   ‚Ä¢ An√°lise hier√°rquica: {self.contador_calculos_rota}")

        print(f"\nüíæ Tamanho dos caches:")
        print(f"   ‚Ä¢ Cache risco: {len(self.cache_risco_propagacao)} entradas")
        print(f"   ‚Ä¢ Cache fibonacci: {len(self.cache_fibonacci_recursos)} entradas")
        print(f"   ‚Ä¢ Cache pot√™ncia: {len(self.cache_potencia_fogo)} entradas")
        print(f"   ‚Ä¢ Cache hierarquia: {len(self.cache_rota_otima)} entradas")

        total_cache = (len(self.cache_risco_propagacao) + len(self.cache_fibonacci_recursos) +
                      len(self.cache_potencia_fogo) + len(self.cache_rota_otima))
        total_calculos = (self.contador_calculos_risco + self.contador_calculos_fibonacci +
                         self.contador_calculos_potencia + self.contador_calculos_rota)

        if total_calculos > 0:
            eficiencia = ((total_cache / (total_calculos + total_cache)) * 100)
            print(f"\n‚ö° Efici√™ncia da memoriza√ß√£o: {eficiencia:.1f}%")
            print(f"   (Cache hits vs c√°lculos realizados)")

# cria sistema de an√°lise inteligente
print("üöÄ INICIALIZANDO SISTEMA DE AN√ÅLISE INTELIGENTE...")
analise = AnaliseInteligente()

# ================================================================
#                    DEMONSTRA√á√ÉO DAS FUN√á√ïES RECURSIVAS
# ================================================================

print("\n" + "="*80)
print("üß™ DEMONSTRA√á√ÉO DE FUN√á√ïES RECURSIVAS COM memoriza√ß√£o")
print("="*80)

print("\nüî• 1. TESTANDO C√ÅLCULO DE RISCO DE PROPAGA√á√ÉO (COM memoriza√ß√£o):")
print("-" * 60)

# primeira chamada - ser√° calculada
municipios_teste = ["Colniza", "Barreiras", "Formoso do Araguaia"]
for municipio in municipios_teste:
    print(f"\nüìç Analisando {municipio}:")
    risco_7_dias = analise.calcular_risco_propagacao(municipio, 7, vento=8)
    print(f"   Risco em 7 dias: {risco_7_dias:.2f}")

print(f"\nüîÑ TESTANDO CACHE - Recalculando mesmos valores:")
# segunda chamada - deve usar cache
for municipio in municipios_teste:
    print(f"\nüìç Recalculando {municipio}:")
    risco_7_dias = analise.calcular_risco_propagacao(municipio, 7, vento=8)
    print(f"   Risco em 7 dias: {risco_7_dias:.2f} (deve ter usado cache!)")

print(f"\nüöÅ 2. TESTANDO FIBONACCI PARA RECURSOS (COM memoriza√ß√£o):")
print("-" * 60)

niveis_emergencia = [10, 12, 10, 15, 12]  # propositalmente repetidos
for nivel in niveis_emergencia:
    print(f"\nüìä N√≠vel de emerg√™ncia {nivel}:")
    recursos = analise.calcular_recursos_fibonacci(nivel)
    print(f"   Recursos necess√°rios: {recursos} unidades")

print(f"\n‚ö° 3. TESTANDO POT√äNCIA DE COMBATE (COM memoriza√ß√£o):")
print("-" * 60)

potencias_teste = [(3, 4), (2, 8), (3, 4), (5, 3), (2, 8)]  # propositalmente repetidos
for base, exp in potencias_teste:
    print(f"\nüî¢ Calculando {base}^{exp}:")
    potencia = analise.calcular_potencia_combate(base, exp)
    print(f"   Resultado: {potencia}")

print(f"\nüå≥ 4. TESTANDO AN√ÅLISE HIER√ÅRQUICA (COM memoriza√ß√£o):")
print("-" * 60)

regioes_teste = ["Centro_Oeste", "Nordeste", "Centro_Oeste"]  # propositalmente repetido
for regiao in regioes_teste:
    print(f"\nüó∫Ô∏è Analisando hierarquia de {regiao}:")
    analise_resultado = analise.analisar_hierarquia_regioes(regiao, 2)
    print(f"   Criticidade: {analise_resultado['criticidade']}")
    print(f"   Sub-regi√µes analisadas: {len(analise_resultado.get('sub_regioes', []))}")

# mostra as estat√≠sticas finais
print(f"\n" + "="*80)
print("üìä RESULTADOS DA MEMOIZA√á√ÉO")
print("="*80)

analise.mostrar_estatisticas_cache()
```
